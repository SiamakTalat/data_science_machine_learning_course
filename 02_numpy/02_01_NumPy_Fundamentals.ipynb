## Part 1 — What is NumPy?

NumPy (Numerical Python) is the foundational library for numerical computing in Python. It provides:
- A fast, fixed-type **n-dimensional array** (`ndarray`)
- Vectorised operations — no Python loops needed for element-wise math
- The backbone for Pandas, Scikit-learn, Matplotlib, and most of the data science ecosystem

Two key array forms:
- **Vector** → one-dimensional array (1D)
- **Matrix** → two-dimensional array (2D), even if it has a single row or column

```python
import numpy as np

print(f"NumPy version: {np.__version__}")
```

> `np` is the universal alias for NumPy — use it consistently throughout this course.

## Part 2 — Creating Arrays from Python Lists

The most direct way to create a NumPy array is by passing a Python list to `np.array()`.

```python
# 1D array from a list
my_list = [1, 2, 3]
arr_1d = np.array(my_list)
arr_1d
```

```python
# 2D array from a list of lists
my_mat = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

arr_2d = np.array(my_mat)
arr_2d
```

**Dimensionality rule:** the number of bracket nesting levels tells you the number of dimensions:
- one bracket level → 1D
- two bracket levels → 2D

## Part 3 — Built-in Array Generation Functions

NumPy provides efficient functions to generate arrays without needing a pre-built Python list.

```python
# arange: like range(), but returns an ndarray
print(np.arange(0, 10))      # 0 to 9
print(np.arange(0, 11, 2))   # even numbers: step of 2
```

```python
# zeros and ones: arrays filled with a constant
print(np.zeros(3))         # 1D
print(np.zeros((2, 3)))    # 2D — dimensions as a tuple (rows, cols)
print(np.ones(4))
print(np.ones((3, 4)))
```

```python
# linspace: N evenly spaced points between start and stop (both inclusive)
print(np.linspace(0, 5, 10))    # 10 points
print(np.linspace(0, 1, 5))     # 5 points
```

**Key distinction:**

| Function | Third argument means |
|----------|---------------------|
| `arange(start, stop, step)` | step size |
| `linspace(start, stop, n)` | number of points |

```python
# eye: identity matrix (1s on diagonal, 0s elsewhere)
np.eye(4)
```

## Part 4 — Random Array Generation

NumPy's `random` module generates arrays from common probability distributions.

```python
# rand: uniform distribution [0, 1)
print(np.random.rand(5))        # 1D
print()
print(np.random.rand(3, 3))     # 2D — arguments are positional, NOT a tuple
```

```python
# randn: standard normal distribution (mean=0, std=1)
print(np.random.randn(4))       # can be negative
print()
print(np.random.randn(3, 3))
```

```python
# randint: random integers in [low, high)  — high is excluded
print(np.random.randint(1, 100))        # single integer
print(np.random.randint(1, 100, 10))    # array of 10 integers
```

## Part 5 — Array Attributes and Reshaping

```python
arr = np.arange(25)
rand_arr = np.random.randint(0, 50, 10)

print("arr:     ", arr)
print("rand_arr:", rand_arr)
```

```python
# reshape: change dimensions without changing data
# total elements must stay the same: 25 -> 5x5 works, 5x10 does not
arr.reshape(5, 5)
```

```python
# max/min return values; argmax/argmin return index positions
print("max value: ", rand_arr.max())
print("min value: ", rand_arr.min())
print("index of max:", rand_arr.argmax())
print("index of min:", rand_arr.argmin())
```

```python
# shape is an attribute (no parentheses), dtype tells you the stored type
arr2 = arr.reshape(5, 5)
print("shape:", arr2.shape)
print("dtype:", arr2.dtype)
```

## Part 6 — Indexing and Slicing (1D)

NumPy indexing mirrors Python list slicing — start included, stop excluded.

```python
arr = np.arange(0, 11)
print(arr)

print(arr[8])      # single element at index 8
print(arr[1:5])    # elements at index 1, 2, 3, 4
print(arr[:6])     # first 6 elements
print(arr[5:])     # from index 5 onward (the 6th element)
```

> **Note:** `arr[5:]` starts at **index 5** (the 6th element). Indexing always starts at zero.

```python
# Broadcasting: assign a scalar to a slice — all selected elements change
arr[0:5] = 100
arr
```

---

## Part 7 — Copy vs View

A NumPy **slice is a view** — it shares memory with the original array. Modifying the slice modifies the original.

```python
arr = np.arange(0, 11)
slice_of_arr = arr[0:6]    # this is a VIEW, not a copy

slice_of_arr[:] = 99

print("slice:   ", slice_of_arr)
print("original:", arr)       # original is also changed!
```

```python
# Use .copy() to get an independent array
arr = np.arange(0, 11)
arr_copy = arr.copy()

arr_copy[:] = 100

print("copy:    ", arr_copy)
print("original:", arr)       # original is unchanged
```

> **Rule:** Use `.copy()` whenever you need to work on a subset without affecting the source array.

## Part 8 — 2D Array Indexing

For 2D arrays, use `[row, col]` notation. Both double-bracket and comma styles work, but comma is preferred.

```python
arr_2d = np.array([
    [ 5, 10, 15],
    [20, 25, 30],
    [35, 40, 45]
])

arr_2d
```

```python
# Double-bracket style (verbose)
print(arr_2d[0][0])   # 5

# Preferred comma style: arr_2d[row, col]
print(arr_2d[1, 2])   # row 1, col 2 -> 30
print(arr_2d[2, 0])   # row 2, col 0 -> 35
```

```python
# Slicing submatrices: [row_slice, col_slice]
print(arr_2d[:2, 1:])     # first 2 rows, columns from index 1 onward
print()
print(arr_2d[1:, :2])     # rows from index 1 onward, first 2 columns
```

## Part 9 — Conditional Selection

Comparing an array with a scalar returns a boolean array. That boolean array can be used to filter values.

```python
arr = np.arange(1, 11)
print(arr)
print(arr > 5)           # boolean array
```

```python
# Use the boolean array as an index filter
bool_arr = arr > 5
print(arr[bool_arr])    # only elements where condition is True

# Inline (most common style)
print(arr[arr > 5])
print(arr[arr < 3])
```

> **This pattern is critical.** It reappears in Pandas as the primary row-filtering mechanism.

## Part 10 — Indexing Practice: Submatrix Selection

Work through selecting specific regions of a larger matrix.

```python
mat = np.arange(1, 26).reshape(5, 5)
mat
```

```python
# Select the bottom-right 3x4 block (rows 2+, cols 1+)
mat[2:, 1:]
```

```python
# Last row only
mat[4]
```

```python
# Last two rows
mat[3:]
```

## Part 11 — Array Arithmetic

Operations are applied **elementwise** — no loops required.

```python
arr = np.arange(0, 11)

print("arr + arr:", arr + arr)
print("arr - arr:", arr - arr)
print("arr * arr:", arr * arr)
```

```python
# Scalar broadcasting: the scalar is applied to every element
print(arr + 100)
print(arr * 2)
print(arr ** 2)
```

```python
# NumPy handles edge cases with warnings, not hard errors
import warnings
warnings.filterwarnings('ignore')   # suppress for clean output

print(arr / arr)    # 0/0 -> nan, others -> 1.0
print(1 / arr)      # 1/0 -> inf
```

**NumPy edge-case behavior:**
- `0 / 0` → `nan` (Not a Number)
- `1 / 0` → `inf` (infinity)

This differs from plain Python, which raises a `ZeroDivisionError`.

## Part 12 — Universal Functions (ufuncs)

Universal functions operate **elementwise** on arrays — they are NumPy's equivalent of math library functions.

```python
arr = np.arange(1, 11)   # avoid 0 for log/sqrt

print("sqrt:", np.sqrt(arr))
print("exp: ", np.exp(arr[:3]))   # truncated for readability
print("log: ", np.log(arr))
print("sin: ", np.sin(arr).round(3))
print("max: ", np.max(arr))
```

```python
# Shortcut: import a specific function directly
from numpy.random import randint

randint(1, 10)   # no np.random. prefix needed
```

## Part 13 — Aggregate Operations on Matrices

`axis=0` → collapse down rows (column-wise result)  
`axis=1` → collapse across columns (row-wise result)

```python
mat = np.arange(1, 26).reshape(5, 5)

print("Total sum:", mat.sum())
print("Std dev:  ", mat.std().round(3))
print("Col sums (axis=0):", mat.sum(axis=0))   # one result per column
print("Row sums (axis=1):", mat.sum(axis=1))   # one result per row
```

## Key Takeaways

- NumPy arrays are faster and more memory-efficient than Python lists for numerical data.
- Use `np.array()`, `arange()`, `zeros()`, `ones()`, `linspace()`, and `eye()` to create arrays.
- `arange(start, stop, step)` vs `linspace(start, stop, n_points)` — know the difference.
- Slices are **views**: changing a slice changes the original. Use `.copy()` to avoid this.
- Conditional selection — `arr[arr > 5]` — is the most important filtering pattern in NumPy and Pandas.
- All arithmetic and universal functions are **vectorised** — no loops needed.
- `axis=0` operates down rows (column results); `axis=1` operates across columns (row results).
