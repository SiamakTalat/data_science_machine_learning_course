## Part 4 — Feature Engineering

Create a new column called `Reason` by extracting the first part of the `title` column (e.g., 'EMS', 'Fire', 'Traffic').

```python
df['Reason'] = df['title'].apply(lambda x: x.split(':')[0])
```

```python
df['Reason'].value_counts()
```

---
## Part 5 — Countplots

Visualize the distribution of 911 calls across different reasons using a Seaborn countplot.

```python
sns.countplot(x='Reason', data=df)
```

---
## Part 6 — Datetime Processing

Convert the `timeStamp` column into actual Python datetime objects to allow for temporal analysis.

```python
df['timeStamp'] = pd.to_datetime(df['timeStamp'])
```

Extract the **Hour**, **Month**, and **Day of Week** from the new datetime column.

```python
df['Hour'] = df['timeStamp'].apply(lambda x: x.hour)df['Month'] = df['timeStamp'].apply(lambda x: x.month)df['Day of Week'] = df['timeStamp'].apply(lambda x: x.dayofweek)
```

---
## Part 7 — Day Mapping

```python
dmap = {
    0:'Mon',
    1:'Tue',
    2:'Wed',
    3:'Thu',
    4:'Fri',
    5:'Sat',
    6:'Sun'
}
```

```python
df['Day of Week'] = df['Day of Week'].map(dmap)
```

---
## Part 8 — Countplots by Day and Month

```python
sns.countplot(x='Day of Week', data=df, hue='Reason')
```

```python
sns.countplot(x='Month', data=df, hue='Reason')
```

### Reflection
- Notice how certain months (e.g., 9, 10, 11) might appear missing in the plots.
- This observation motivates the use of line plots to track trends over time.

---
## Part 9 — Monthly Trends

Create a grouped count of calls per month to identify long-term trends.

```python
byMonth = df.groupby('Month').count()['lat'].reset_index()
```

```python
plt.figure(figsize=(12,6))sns.lineplot(x='Month', y='lat', data=byMonth)
```

---
## Part 10 — Daily Trends

Create a column for the date only (removing time) to observe daily call volumes.

```python
df['Date'] = df['timeStamp'].apply(lambda x: x.date())
```

```python
df.groupby('Date').count()['lat'].plot()plt.tight_layout()
```

---
## Part 11 — Trends by Reason

```python
df[df['Reason']=='Traffic'].groupby('Date').count()['lat'].plot()plt.title('Traffic')
```

```python
df[df['Reason']=='Fire'].groupby('Date').count()['lat'].plot()plt.title('Fire')
```

```python
df[df['Reason']=='EMS'].groupby('Date').count()['lat'].plot()plt.title('EMS')
```

---
## Part 12 — Heatmaps

Transform the data into a matrix format (Day of Week vs. Hour) to visualize peak calling hours.

```python
dayHour = df.groupby(by=['Day of Week','Hour']).count()['Reason'].unstack()
```

```python
plt.figure(figsize=(12,6))sns.heatmap(dayHour, cmap='viridis')
```

```python
sns.clustermap(dayHour, cmap='viridis')
```

---## Part 13 — Monthly Heatmaps

```python
dayMonth = df.groupby(by=['Day of Week','Month']).count()['Reason'].unstack()
```

```python
plt.figure(figsize=(12,6))sns.heatmap(dayMonth, cmap='viridis')
```

```python
sns.clustermap(dayMonth, cmap='viridis')
```
