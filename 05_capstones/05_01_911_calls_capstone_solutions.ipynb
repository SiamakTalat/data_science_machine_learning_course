# 13 — Solutions: 911 Calls Capstone Project

This notebook contains the complete solutions for the 911 Calls Capstone Project. Each cell corresponds to a specific task or step in the original project notebook.

## Part 1 — Setup

```python
import numpy as npimport pandas as pdimport matplotlib.pyplot as pltimport seaborn as sns%matplotlib inline
```

## Part 2 — Load Dataset

```python
df = pd.read_csv('911.csv')
```

## Part 3 — Basic EDA Solutions

```python
df['zip'].value_counts().head(5)df['twp'].value_counts().head(5)df['title'].nunique()
```

## Part 4 — Feature Engineering Solutions

```python
df['Reason'] = df['title'].apply(lambda x: x.split(':')[0])df['Reason'].value_counts()
```

## Part 5 — Countplots Solutions

```python
sns.countplot(x='Reason', data=df)
```

## Part 6 — Datetime Processing Solutions

```python
df['timeStamp'] = pd.to_datetime(df['timeStamp'])df['Hour'] = df['timeStamp'].apply(lambda x: x.hour)df['Month'] = df['timeStamp'].apply(lambda x: x.month)df['Day of Week'] = df['timeStamp'].apply(lambda x: x.dayofweek)
```

## Part 7 — Day Mapping Solutions

```python
dmap = {
    0:'Mon',
    1:'Tue',
    2:'Wed',
    3:'Thu',
    4:'Fri',
    5:'Sat',
    6:'Sun'
}df['Day of Week'] = df['Day of Week'].map(dmap)
```

## Part 8 — Countplots by Day and Month Solutions

```python
sns.countplot(x='Day of Week', data=df, hue='Reason')sns.countplot(x='Month', data=df, hue='Reason')
```

## Part 9 — Monthly Trends Solutions

```python
byMonth = df.groupby('Month').count()byMonth['lat'].plot()sns.lmplot(x='Month', y='lat', data=byMonth.reset_index())
```

## Part 10 — Daily Trends Solutions

```python
df['Date'] = df['timeStamp'].apply(lambda x: x.date())df.groupby('Date').count()['lat'].plot()plt.tight_layout()
```

## Part 11 — Trends by Reason Solutions

```python
plt.figure(figsize=(12,6))df[df['Reason']=='Traffic'].groupby('Date').count()['lat'].plot()plt.title('Traffic')
```

```python
df[df['Reason']=='Fire'].groupby('Date').count()['lat'].plot()plt.title('Fire')
```

```python
df[df['Reason']=='EMS'].groupby('Date').count()['lat'].plot()plt.title('EMS')
```

## Part 12 — Heatmaps Solutions

```python
dayHour = df.groupby(by=['Day of Week','Hour']).count()['Reason'].unstack()sns.heatmap(dayHour, cmap='viridis')
```

```python
sns.clustermap(dayHour, cmap='viridis')
```

## Part 13 — Monthly Heatmaps Solutions

```python
dayMonth = df.groupby(by=['Day of Week','Month']).count()['Reason'].unstack()plt.figure(figsize=(12,6))sns.heatmap(dayMonth, cmap='viridis')
```

```python
sns.clustermap(dayMonth, cmap='viridis')
```
