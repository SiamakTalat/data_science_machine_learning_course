## 1. What are Branches?

A **branch** is a lightweight, movable pointer to a specific commit. The default branch is named `main` (or `master` in older repos).

Creating a branch gives you an isolated workspace where you can experiment, fix a bug, or develop a new feature — all without touching the stable code on `main`.

```
main:      A ── B ── C
                     ↑
feature:             C ── D ── E
```

| Concept | Meaning |
|---------|--------|
| `HEAD` | Pointer to the current branch (your "you are here" marker) |
| `main` | The primary production/stable branch |
| Feature branch | A short-lived branch for one piece of work |
| Merge | Combining one branch's changes back into another |

> **Key idea:** Branches are just pointers — creating one is instantaneous and uses almost no disk space.

---

## 2. Creating and Switching Branches

Two modern commands cover all branch creation and switching: `git branch` to create and `git switch` to move. The older `git checkout -b` form still works and is common in documentation.

````python
# Creating and switching branches
#
#  Create a new branch (does NOT switch to it)
#   git branch feature/eda
#
#  Switch to an existing branch
#   git switch feature/eda
#
#  Create AND switch in one step (most common)
#   git switch -c feature/eda
#
#  Older equivalent (still widely used)
#   git checkout -b feature/eda
#
#  Switch back to main
#   git switch main
#
#  Switch to the previous branch (like cd -)
#   git switch -

print("git switch -c <name> is the modern way to create and enter a new branch.")
````

---

## 3. Listing Branches

`git branch` with no arguments lists all local branches. The branch with `*` is your current one (where `HEAD` points).

````python
# Listing branches
#
#  List local branches (* marks the current branch)
#   git branch
#   → * main
#   →   feature/eda
#   →   bugfix/null-values
#
#  List remote-tracking branches
#   git branch -r
#
#  List all branches (local + remote)
#   git branch -a
#
#  List branches with their last commit message
#   git branch -v
#
#  List branches already merged into main (safe to delete)
#   git branch --merged main

print("git branch -v shows the last commit on each branch at a glance.")
````

---

## 4. Making Commits on a Branch

Once you are on a feature branch, all `git add` and `git commit` operations work exactly as on `main` — the difference is that `main` remains untouched until you explicitly merge.

````python
# Typical feature branch workflow
#
# 1. Start a new branch from main
#   git switch main
#   git switch -c feature/clean-data
#
# 2. Do your work — edit files, run cells, generate outputs
#
# 3. Stage and commit as normal
#   git add data_cleaning.py
#   git commit -m "Add missing value imputation step"
#
# 4. Continue committing as you work
#   git add data_cleaning.py
#   git commit -m "Add outlier removal for price column"
#
# 5. Check where each branch is
#   git log --oneline --graph --all

print("main stays frozen while you commit freely on the feature branch.")
````

---

## 5. Merging Branches — `git merge`

When the work on a feature branch is complete, merge it back into `main`. **Always switch to the target branch first** — you merge *into* the current branch.

````python
# Merging a feature branch into main
#
# 1. Switch to the branch you want to merge INTO
#   git switch main
#
# 2. Merge the feature branch
#   git merge feature/clean-data
#
# 3. If successful:
#   → Fast-forward or merge commit created
#   → main now includes all commits from feature/clean-data
#
# 4. Delete the feature branch (optional, but keeps the repo tidy)
#   git branch -d feature/clean-data
#
# To create a merge commit even when a fast-forward is possible:
#   git merge --no-ff feature/clean-data
#   (useful to preserve that a branch existed in the history graph)

print("Switch to main FIRST, then merge the feature branch into it.")
````

## 6. Fast-forward vs Three-way Merge

Git chooses the merge strategy automatically based on the commit history.

**Fast-forward** — `main` has not moved since the branch was created. Git simply moves the `main` pointer forward. No extra merge commit is created.

```
Before:   main: A──B       feature: A──B──C──D
After FF: main: A──B──C──D
```

**Three-way merge** — `main` has new commits that are not in the feature branch. Git creates a new **merge commit** with two parents.

```
Before:   main: A──B──E         feature: A──B──C──D
After:    main: A──B──E──M (merge commit, parents E and D)
                    ↗         ↗
                   C──D
```

````python
# Inspecting which type of merge occurred
#
#  After a fast-forward merge:
#   git log --oneline
#   → d1e2f3a Add outlier removal          ← these were on the feature branch
#   → b4c5d6e Add missing value imputation
#   → a1b2c3d Initial commit
#   (no merge commit visible)
#
#  After a three-way merge:
#   git log --oneline --graph
#   → *   9f8e7d6 Merge branch 'feature/clean-data'
#   → |\  
#   → | * d1e2f3a Add outlier removal
#   → | * b4c5d6e Add missing value imputation
#   → * | e4f5a6b Update README on main
#   → |/
#   → * a1b2c3d Initial commit

print("Use --graph with git log to see the merge topology clearly.")
````

## 7. Merge Conflicts

A **merge conflict** occurs when two branches have changed the same part of the same file in incompatible ways. Git cannot decide which version to keep — it stops and asks you to resolve it.

Conflict markers in the file:
```
<<<<<<< HEAD
result = df.dropna()
=======
result = df.fillna(0)
>>>>>>> feature/clean-data
```

Everything between `<<<<<<<` and `=======` is the current branch version; between `=======` and `>>>>>>>` is the incoming branch version.

````python
# Resolving a merge conflict — step by step
#
# 1. Run git merge — it reports a conflict
#   git merge feature/clean-data
#   → CONFLICT (content): Merge conflict in analysis.py
#   → Automatic merge failed; fix conflicts and then commit.
#
# 2. Open the conflicted file — look for <<<<<<<, =======, >>>>>>>
#   Edit the file to keep the version you want (delete the markers too)
#
# 3. Stage the resolved file
#   git add analysis.py
#
# 4. Complete the merge
#   git commit
#   (Git pre-fills a merge commit message — accept it or edit it)
#
# To abort the merge and return to the state before git merge:
#   git merge --abort

print("After resolving conflict markers: git add the file, then git commit to finish.")
````

## 8. Deleting and Renaming Branches

Delete a branch once it has been merged — keeping stale branches around makes navigation confusing.

````python
# Deleting branches
#
#  Safe delete (only if branch is already merged)
#   git branch -d feature/clean-data
#
#  Force delete (discards unmerged work — use with care)
#   git branch -D feature/experiment
#
#  Delete a remote branch (on GitHub)
#   git push origin --delete feature/clean-data
#
# Renaming the current branch
#   git branch -m new-name
#
# Renaming any branch
#   git branch -m old-name new-name
#
# Rename the default branch from master to main
#   git branch -m master main

print("-d refuses to delete unmerged branches; use -D only when you know the work is disposable.")
````

## 9. Branch Naming Conventions

Consistent naming makes branches self-documenting. Slashes (`/`) create logical groupings in most Git tools.

| Prefix | Purpose | Example |
|--------|---------|--------|
| `feature/` | New functionality | `feature/customer-churn-model` |
| `bugfix/` | Fix a defect | `bugfix/missing-values-crash` |
| `hotfix/` | Urgent production fix | `hotfix/api-key-leak` |
| `experiment/` | Exploratory / throwaway | `experiment/lstm-approach` |
| `docs/` | Documentation only | `docs/update-readme` |
| `release/` | Release preparation | `release/v1.2.0` |

> Use lowercase with hyphens (not underscores or spaces). Keep names short but descriptive.

## 10. A Simple Branching Strategy for Data Science

Large teams use formal workflows (Git Flow, GitHub Flow). For data science notebooks a simpler model works well:

```
main          → clean, working, reviewed code only
  └─ feature/eda              → exploratory analysis
  └─ feature/feature-engineering  → pipeline changes
  └─ experiment/xgboost       → model experiments
```

**Rules:**
1. Never commit directly to `main` — always use a branch
2. One branch per logical piece of work
3. Merge (or open a pull request) when the work is complete and reviewed
4. Delete the branch after merging

````python
# Complete feature branch lifecycle
#
# 1. Start from a clean main
#   git switch main
#   git pull                         # make sure main is up to date
#
# 2. Create a feature branch
#   git switch -c feature/eda
#
# 3. Work and commit
#   git add eda_notebook.ipynb
#   git commit -m "Add EDA for sales dataset"
#
# 4. Merge back into main
#   git switch main
#   git merge feature/eda
#
# 5. Clean up
#   git branch -d feature/eda

print("Full cycle: branch → work → merge → delete. Repeat.")
````

## Key Takeaways

- A **branch** is a pointer to a commit — creating one is instantaneous and cheap.
- `git switch -c <name>` creates and enters a new branch in one step.
- All commits on a branch leave `main` untouched until you explicitly merge.
- **Fast-forward** merge: no new commit — `main` pointer moves forward. **Three-way** merge: a new merge commit is created with two parents.
- Merge conflicts occur when both branches edited the same lines — resolve by editing the conflict markers, staging, and committing.
- Use `git merge --abort` to cancel an in-progress conflicted merge safely.
- Delete merged branches with `git branch -d` to keep the repo tidy.
