## 1. What is Git?

**Git** is a distributed version control system. It records a complete history of every change made to a project so you can:
- Revert files to any earlier state
- See exactly what changed, when, and who changed it
- Work on multiple features in parallel without breaking the main code
- Collaborate with others safely

| Concept | Meaning |
|---------|--------|
| **Repository (repo)** | The folder Git is tracking — includes `.git/` |
| **Commit** | A saved snapshot of the project at a point in time |
| **Branch** | An independent line of development |
| **Remote** | A copy of the repo hosted elsewhere (e.g. GitHub) |

> **Key idea:** Git is a local tool. GitHub is a hosting platform for Git repositories. You can use Git entirely offline — no internet required.

## 3. Git Configuration

Before your first commit, tell Git who you are. This name and email are stamped into every commit you make.

Configuration is stored in `~/.gitconfig` and applies to all repos on your machine (`--global`).

````python
# Git configuration — run these once in your terminal (not inside a notebook)
#
# Set your name and email (used in every commit)
#   git config --global user.name  "Your Name"
#   git config --global user.email "you@example.com"
#
# Set a default branch name (modern standard is 'main')
#   git config --global init.defaultBranch main
#
# Set VS Code as the default editor
#   git config --global core.editor "code --wait"
#
# View your current configuration
#   git config --list

result = subprocess.run(['git', 'config', '--list'], capture_output=True, text=True)
for line in result.stdout.strip().splitlines()[:8]:
    print(line)
````

## 4. Initialising a Repository

`git init` turns any folder into a Git repository by creating a hidden `.git/` sub-directory. All history and configuration for that repo live inside `.git/`.

The alternative is `git clone <url>` — which copies an existing repository from a remote (covered in Section 50C).

````python
# Typical workflow to start a new project
#
# 1. Create a project folder and move into it
#   mkdir my-project
#   cd my-project
#
# 2. Initialise Git
#   git init
#   → Initialised empty Git repository in /path/to/my-project/.git/
#
# 3. Confirm the .git folder was created
#   ls -a
#   → .  ..  .git
#
# To stop tracking a folder with Git, simply delete the .git/ directory:
#   rm -rf .git

print("git init creates a .git/ folder — this IS the repository.")
print("Delete .git/ to remove Git tracking without deleting your files.")
````

## 5. The Three Areas

Understanding where your files live at each moment is essential for using Git confidently.

```
Working Tree  →  git add  →  Staging Area  →  git commit  →  Repository
(edited files)               (files ready              (permanent history)
                              to be committed)
```

| Area | Also called | Purpose |
|------|------------|--------|
| Working tree | Working directory | Where you edit files |
| Staging area | Index, cache | Files selected for the next commit |
| Repository | `.git/` | The permanent commit history |

> Staging lets you choose **exactly which changes** to include in a commit, even if you have edited multiple files.

---

## 6. Staging Changes — `git add`

`git add` moves changes from the working tree into the staging area. Nothing is saved to history yet — that happens on commit.

````python
# git add — common usage patterns
#
#  Stage a single file
#   git add README.md
#
#  Stage all changes in the current directory
#   git add .
#
#  Stage all .py files
#   git add *.py
#
#  Stage parts of a file interactively (choose individual hunks)
#   git add -p README.md
#
#  Remove a file from staging (un-stage without discarding edits)
#   git restore --staged README.md

print("git add selects changes for the next commit — nothing is permanent yet.")
````

---

## 7. Committing Changes — `git commit`

A **commit** is a permanent snapshot of the staged changes. Each commit gets a unique 40-character SHA hash, an author, a timestamp, and a message.

Write commit messages in the **imperative mood** that completes the sentence *"If applied, this commit will…"*:
- `Add exploratory analysis notebook` ✓
- `Fix missing value handling in feature pipeline` ✓
- `Added stuff` ✗

````python
# Committing — common patterns
#
#  Standard commit (opens editor for the message)
#   git commit
#
#  Commit with an inline message (most common)
#   git commit -m "Add exploratory analysis notebook"
#
#  Stage all tracked files and commit in one step
#   git commit -am "Fix typo in README"
#   Note: -a does NOT include new (untracked) files — use git add for those
#
#  Amend the most recent commit (fix message or add forgotten file)
#   git commit --amend
#   Warning: do not amend commits that have already been pushed

print("Every commit is a permanent, addressable snapshot in history.")
````

## 8. Checking Status and Differences

`git status` tells you what has changed since the last commit and what is staged. Use it constantly — it is the safest way to understand where you are before running any other command.

````python
# git status — understanding the output
#
#  $ git status
#  On branch main
#  Changes to be committed:       ← staged (green in terminal)
#    modified: analysis.py
#  Changes not staged for commit:  ← edited but not staged (red)
#    modified: README.md
#  Untracked files:               ← new files Git doesn't know about
#    data/raw/survey.csv
#
# git diff — show unstaged changes (working tree vs staging area)
#   git diff
#
# git diff --staged — show staged changes (staging area vs last commit)
#   git diff --staged
#
# Compact one-line status
#   git status -s

print("Read git status before every add/commit — it tells you exactly what will be included.")
````

---

## 9. Viewing History — `git log`

`git log` shows the full commit history newest-first. Each entry shows the SHA, author, date, and message.

````python
# git log — useful display formats
#
#  Full default output
#   git log
#
#  Compact one-line per commit
#   git log --oneline
#
#  Compact with branch/merge graph
#   git log --oneline --graph --all
#
#  Show changes introduced by each commit
#   git log -p
#
#  Limit to last N commits
#   git log -5
#
#  Filter by author
#   git log --author="Alice"
#
#  Filter by date range
#   git log --after="2024-01-01" --before="2024-06-01"

print("git log --oneline --graph --all gives the clearest picture of branch history.")
````

## 10. Undoing Changes

Git provides several ways to undo work depending on how far along the change is.

| Situation | Command |
|-----------|--------|
| Discard edits in working tree (not staged) | `git restore <file>` |
| Un-stage a file (keep edits) | `git restore --staged <file>` |
| Move HEAD back, keep changes staged | `git reset --soft HEAD~1` |
| Move HEAD back, keep changes unstaged | `git reset --mixed HEAD~1` |
| Move HEAD back, discard all changes | `git reset --hard HEAD~1` |
| Create a new commit that reverses a commit | `git revert <sha>` |

> `git revert` is the **safe** undo for shared history — it adds a new commit rather than rewriting history.

````python
# Undo cheat sheet (run in terminal)
#
#  Discard unsaved edits to a file (working tree)
#   git restore README.md
#
#  Un-stage a file without losing your edits
#   git restore --staged README.md
#
#  Undo the last commit but keep its changes staged
#   git reset --soft HEAD~1
#
#  Undo the last commit, un-stage changes (most useful for rewording a message)
#   git reset --mixed HEAD~1
#
#  Safely reverse a published commit (creates a new reversal commit)
#   git revert abc1234
#
#  View what HEAD~1 means — inspect the commit before the current one
#   git show HEAD~1

print("Use git revert for undo on shared branches; git reset only on local unpushed commits.")
````

## Key Takeaways

- **Git** is a local version control system — GitHub is a hosting platform built on top of Git.
- Configure `user.name` and `user.email` once with `git config --global` — these are stamped on every commit.
- Changes move through three areas: **working tree → staging area → repository** via `git add` then `git commit`.
- `git status` before every operation — it shows exactly which files are tracked, staged, or modified.
- Write commit messages in the imperative mood: *"Add feature"*, not *"Added feature"*.
- Use `git revert` to undo commits safely on shared branches; use `git reset` only on local unpushed commits.
- `git log --oneline --graph --all` gives the clearest compact view of history across all branches.
