## 1. What is a Remote?

A **remote** is a version of your repository hosted on another computer — typically GitHub, GitLab, or Bitbucket. You can have multiple remotes per repo, but almost all projects use one called **`origin`** by convention.

| Term | Meaning |
|------|---------|
| `origin` | The remote you cloned from (set automatically by `git clone`) |
| `upstream` | The original repo when you work on a fork |
| `origin/main` | Your local copy of origin's `main` at the last fetch |
| Tracking branch | A local branch linked to a remote branch |

> **Key idea:** Your local repo and the remote are independent. Changes do not sync automatically — you push and pull explicitly.

---

## 2. Cloning a Repository

`git clone` creates a full local copy of a remote repository — including all commits, branches, and history. It also sets `origin` to the source URL automatically.

````python
# Cloning a repository from GitHub
#
#  Clone into a folder named after the repo
#   git clone https://github.com/user/repo.git
#
#  Clone into a custom folder name
#   git clone https://github.com/user/repo.git my-folder
#
#  Clone with SSH (requires SSH key set up in GitHub settings)
#   git clone git@github.com:user/repo.git
#
#  After cloning, origin is configured automatically:
#   git remote -v
#   → origin  https://github.com/user/repo.git (fetch)
#   → origin  https://github.com/user/repo.git (push)

print("git clone copies the full repo history and sets origin for you.")
````

---

## 3. Managing Remotes

When you start a repo locally with `git init`, there is no remote configured yet. You add one manually before your first push.

````python
# Managing remotes
#
#  List all configured remotes (with URLs)
#   git remote -v
#
#  Add a remote (standard first step after git init)
#   git remote add origin https://github.com/you/my-project.git
#
#  Rename a remote
#   git remote rename origin backup
#
#  Change a remote's URL (e.g. after renaming a GitHub repo)
#   git remote set-url origin https://github.com/you/renamed-repo.git
#
#  Remove a remote
#   git remote remove origin
#
#  Inspect a remote — shows tracked branches and push/fetch URLs
#   git remote show origin

print("Use git remote -v after any change to verify the URL is correct.")
````

---

## 4. Fetching Changes — `git fetch`

`git fetch` downloads new commits and branches from the remote **without touching your working tree or local branches**. It updates `origin/main` (the remote-tracking branch) so you can review changes before applying them.

````python
# git fetch — download without merging
#
#  Fetch all branches from origin
#   git fetch origin
#
#  Fetch a specific branch
#   git fetch origin main
#
#  After fetching, compare your main with origin/main
#   git diff main origin/main
#
#  See commits on origin/main that you don't have yet
#   git log main..origin/main --oneline
#
#  Then merge if you're happy with the changes
#   git merge origin/main

print("git fetch is always safe — it never changes your working files.")
````

---

## 5. Pulling Changes — `git pull`

`git pull` is shorthand for `git fetch` followed immediately by `git merge`. It updates your current branch to match the remote in one step.

| Command | Effect |
|---------|-------|
| `git fetch` | Download only — your branch is unchanged |
| `git pull` | Download + merge into current branch |
| `git pull --rebase` | Download + rebase instead of merge (cleaner history) |

````python
# git pull — common patterns
#
#  Pull from origin into current branch (most common usage)
#   git pull
#
#  Pull a specific branch from a specific remote
#   git pull origin main
#
#  Pull using rebase instead of merge (avoids merge commits in history)
#   git pull --rebase
#
#  Set rebase as the default pull strategy globally
#   git config --global pull.rebase true
#
# Best practice: always pull before starting new work
#   git switch main
#   git pull             # ensure you're starting from the latest main
#   git switch -c feature/new-thing

print("Pull before branching — start feature work from the latest main.")
````

---

## 6. Pushing Changes — `git push`

`git push` uploads your local commits to the remote. The first time you push a new branch, use `-u` to set the tracking relationship — after that, `git push` alone is enough.

````python
# git push — common patterns
#
#  Push current branch to origin (after -u has been set)
#   git push
#
#  First push of a new branch — sets upstream tracking
#   git push -u origin feature/eda
#   → After this, plain 'git push' works for this branch
#
#  Push to a specific remote and branch
#   git push origin main
#
#  Push all local branches
#   git push --all
#
#  Push tags
#   git push --tags
#
#  Delete a remote branch (e.g. after a PR is merged)
#   git push origin --delete feature/eda
#
# WARNING: Force push overwrites remote history — never force push to main
#   git push --force   # dangerous on shared branches

print("Use -u on the first push to set tracking; plain 'git push' works after that.")
````

---

## 7. Tracking Branches

A **tracking branch** is a local branch linked to a remote branch. Git uses this link to know where to push and pull without specifying the remote and branch name each time.

`origin/main` is a **remote-tracking branch** — a read-only local reference to where `main` was on origin at the last fetch.

````python
# Tracking branches — inspection commands
#
#  Show which remote branch each local branch tracks
#   git branch -vv
#   → * main    a1b2c3d [origin/main] Update README
#   →   feature  d4e5f6g [origin/feature/eda: ahead 2] Add visualisations
#
#  'ahead 2' means you have 2 local commits not yet pushed
#  'behind 3' means origin has 3 commits you haven't pulled
#
#  Set a tracking relationship for an existing branch
#   git branch --set-upstream-to=origin/main main
#
#  Create a local branch from a remote branch (and track it)
#   git switch --track origin/feature/eda

print("git branch -vv shows ahead/behind counts for every tracked branch.")
````

## 8. The Fork → Clone → Branch → PR Workflow

This is the standard open-source contribution workflow. It lets you contribute to any repo without needing write access to the original.

```
Step 1: Fork   → GitHub makes your copy of the repo under your account
Step 2: Clone  → git clone your fork to your local machine
Step 3: Branch → git switch -c feature/my-change
Step 4: Commit → make changes, git add, git commit
Step 5: Push   → git push -u origin feature/my-change
Step 6: PR     → open a Pull Request on GitHub from your fork to the original
```

````python
# Full contribution workflow — terminal commands
#
# Step 1: Fork on GitHub (click Fork button on the repo page)
#
# Step 2: Clone YOUR fork
#   git clone https://github.com/YOUR-USERNAME/repo.git
#   cd repo
#
# Step 3: Add the original as 'upstream'
#   git remote add upstream https://github.com/ORIGINAL-OWNER/repo.git
#   git remote -v    # verify: origin = your fork, upstream = original
#
# Step 4: Create a branch
#   git switch -c feature/fix-typo
#
# Step 5: Make changes and commit
#   git add README.md
#   git commit -m "Fix typo in installation instructions"
#
# Step 6: Push to YOUR fork
#   git push -u origin feature/fix-typo
#
# Step 7: Open a PR on GitHub — from your fork's feature branch → upstream main

print("Always branch from a fresh main — never commit directly to your fork's main.")
````

---

## 9. Syncing a Fork with Upstream

After others merge PRs into the original repo, your fork's `main` falls behind. Sync it regularly to avoid large conflicts when you start the next contribution.

````python
# Syncing your fork — keep it up to date with the original
#
# 1. Fetch all new commits from upstream
#   git fetch upstream
#
# 2. Switch to your local main
#   git switch main
#
# 3. Merge upstream/main into your local main
#   git merge upstream/main
#   (or use rebase: git rebase upstream/main)
#
# 4. Push the updated main to YOUR fork on GitHub
#   git push origin main
#
# Shortcut — GitHub also provides a 'Sync fork' button on your fork's page

print("Sync your fork before starting any new feature branch.")
````

## 10. Pull Requests

A **Pull Request (PR)** is a GitHub feature (not a Git command) that proposes merging one branch into another. It adds a review and discussion layer before the merge happens.

| PR Element | Purpose |
|------------|--------|
| **Title** | One-line summary of what the PR does |
| **Description** | What changed, why, and how to test it |
| **Reviewers** | People asked to review the code |
| **Checks** | CI/CD tests that run automatically |
| **Files changed** | Diff view of every modified file |
| **Merge button** | Merges the branch when approved |

> After a PR is merged on GitHub, sync locally: `git switch main && git pull && git branch -d feature/...`

````python
# Clean up after a PR is merged on GitHub
#
# 1. Switch to local main
#   git switch main
#
# 2. Pull the merged changes (your PR is now part of origin/main)
#   git pull
#
# 3. Delete the local feature branch (it's merged, no longer needed)
#   git branch -d feature/fix-typo
#
# 4. Optionally delete the remote branch if GitHub didn't do it automatically
#   git push origin --delete feature/fix-typo

print("After every merged PR: pull main, delete the branch, start fresh.")
````

## Key Takeaways

- A **remote** is a named URL for another copy of the repo. `origin` is the name Git assigns automatically when you clone.
- `git clone` copies the full history and configures `origin` for you; `git remote add origin <url>` is used after `git init`.
- `git fetch` is always safe — it downloads commits without changing your working files or local branches.
- `git pull` = `git fetch` + `git merge`. Use `git pull --rebase` for a cleaner linear history.
- Use `git push -u origin <branch>` for the first push of a new branch; plain `git push` works after that.
- The standard open-source contribution cycle: **fork → clone → branch → commit → push → pull request**.
- After a PR is merged, sync locally: `git pull` on `main`, then delete the feature branch.
