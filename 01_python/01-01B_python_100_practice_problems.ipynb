## Problem 21: Generate first n prime numbers


```python
def is_prime(num):
    if num < 2:
        return False
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            return False
    return True

def prime_generator(n):
    count = 0
    num = 2
    while count < n:
        if is_prime(num):
            yield num
            count += 1
        num += 1

for p in prime_generator(10):
    print(p, end=" ")
```

## Problem 61: Merge two sorted lists


```python
a = [1, 3, 5, 7]
b = [2, 4, 6, 8]

i = 0
j = 0
merged = []

while i < len(a) and j < len(b):
    if a[i] < b[j]:
        merged.append(a[i])
        i += 1
    else:
        merged.append(b[j])
        j += 1

merged.extend(a[i:])
merged.extend(b[j:])

print(merged)
```

## Problem 64: Binary search


```python
nums = [2, 4, 6, 8, 10, 12, 14, 16]
target = 10

left = 0
right = len(nums) - 1
found = -1

while left <= right:
    mid = (left + right) // 2
    if nums[mid] == target:
        found = mid
        break
    elif nums[mid] < target:
        left = mid + 1
    else:
        right = mid - 1

print(found)
```

## Problem 87: Generate Pascal's triangle


```python
n = 5
triangle = []

for i in range(n):
    row = [1] * (i + 1)
    for j in range(1, i):
        row[j] = triangle[i - 1][j - 1] + triangle[i - 1][j]
    triangle.append(row)

for row in triangle:
    print(row)
```

## Problem 88: Find the length of the longest substring without repeating characters


```python
s = "abcabcbb"
start = 0
seen = {}
max_len = 0

for i, ch in enumerate(s):
    if ch in seen and seen[ch] >= start:
        start = seen[ch] + 1
    seen[ch] = i
    max_len = max(max_len, i - start + 1)

print(max_len)
```

## Problem 91: Check whether two words differ by one edit


```python
a = "pale"
b = "ple"

if abs(len(a) - len(b)) > 1:
    print(False)
else:
    if len(a) < len(b):
        a, b = b, a

    i = j = 0
    found = False
    one_edit = True

    while i < len(a) and j < len(b):
        if a[i] != b[j]:
            if found:
                one_edit = False
                break
            found = True
            if len(a) == len(b):
                j += 1
        else:
            j += 1
        i += 1

    print(one_edit)
```

## Problem 92: Implement selection sort


```python
nums = [64, 25, 12, 22, 11]

for i in range(len(nums)):
    min_idx = i
    for j in range(i + 1, len(nums)):
        if nums[j] < nums[min_idx]:
            min_idx = j
    nums[i], nums[min_idx] = nums[min_idx], nums[i]

print(nums)
```

## Problem 93: Implement insertion sort


```python
nums = [12, 11, 13, 5, 6]

for i in range(1, len(nums)):
    key = nums[i]
    j = i - 1
    while j >= 0 and nums[j] > key:
        nums[j + 1] = nums[j]
        j -= 1
    nums[j + 1] = key

print(nums)
```
