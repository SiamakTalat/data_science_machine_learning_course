## Part 5 — String Indexing and Slicing

Strings are indexed sequences. Index starts at 0. Slicing uses `[start:end]` — start is **included**, end is **excluded**.

```python
s = "abcdefghijk"

print(s[0])     # first character: 'a'
print(s[:3])    # first 3 characters: 'abc'
print(s[3:6])   # characters at index 3, 4, 5: 'def'
print(s[-1])    # last character: 'k'
```

```python
str2 = "Hi friends, This is a pyhon class"
str3 = str2[-1]
str4 = str2[-8]
str5 = str2[:-1]
str6 = str2[:-5]
str7 = str2[3:-28]

print(str3)
print(str4)
print(str5)
print(str6)
print(str7)
```

## Part 9 — Tuples and Sets

**Tuple:** ordered, **immutable** sequence. Once created, elements cannot be changed.

**Set:** unordered collection of **unique** elements.

```python
# Tuple — immutable
t = (1, 2, 3)
print(t[0])          # indexing works
# t[0] = 99         # ← this would raise a TypeError
```

```python
# Set — unique elements only, duplicates removed automatically
s = set([1, 1, 2, 2, 3, 3])
print(s)   # {1, 2, 3}
```

| Type | Ordered | Mutable | Unique |
|------|---------|---------|--------|
| list | yes | yes | no |
| tuple | yes | **no** | no |
| set | no | yes | **yes** |

## Part 10 — Comparison and Logic Operators

```python
# Comparison operators return True or False
print(1 < 2)    # True
print(1 == 1)   # True  (equality)
print(1 != 3)   # True  (not equal)
print(2 >= 5)   # False
```

## Part 13 — While Loops

`while` loops run until a condition becomes False.

```python
i = 1

while i < 5:
    print(i)
    i += 1   # IMPORTANT: always update the loop variable
```

## Part 15 — List Comprehension

List comprehension is a concise way to build a list in one line. It replaces the loop-and-append pattern.

```python
# Traditional approach — loop and append
x = [1, 2, 3, 4]
out = []

for num in x:
    out.append(num ** 2)

out
```

```python
# Pythonic approach — list comprehension
[num ** 2 for num in x]
```

```python
# With a filter condition
[num for num in x if num % 2 == 0]   # even numbers only
```

> **Pattern:** `[expression for item in iterable if condition]`

## Part 17 — Docstrings

A docstring is a string placed immediately inside a function definition. It documents what the function does.

## Part 18 — Lambda Expressions

## Part 19 — Map and Filter

`map()` applies a function to every element. `filter()` keeps only elements where the function returns `True`.

```python
seq = [1, 2, 3, 4, 5]

# map: double every element
list(map(lambda x: x * 2, seq))
```

```python
# filter: keep only even elements
list(filter(lambda x: x % 2 == 0, seq))
```

| Function | Purpose | Returns |
|----------|---------|--------|
| `map(fn, seq)` | Apply `fn` to each element | New sequence, same length |
| `filter(fn, seq)` | Keep elements where `fn` returns True | Subset of original |

> Both return lazy iterators — wrap with `list()` to see results.

{yrs:>5.2f} is a Python f-string format specifier that controls number formatting and alignment.

yrs → the variable\
5 → minimum width\
.2 → 2 decimal places\
f → float (fixed-point) format\
\> → right align\
< → left align\
^ → center align

```python
yrs = 3.4567

f"{yrs:>6.2f}"   # '  3.46'  (right-aligned)
f"{yrs:<6.2f}"   # '3.46  '  (left-aligned)
f"{yrs:^6.2f}"   # ' 3.46 '  (centered)
```

```python
num = 12345678  
print(f"{num:5.2f}")  # 5 = minimum width, not maximum
```

f → float → {:5.2f}\
g → general → {:5.2g}

```python
num = 12345678
print(f"{num:5.2g}")
```

## Part 21 — Tuple Unpacking

Tuples can be unpacked into named variables in a single line — very common in loops.

```python
# Tuple unpacking in a for loop
pairs = [(1, 2), (3, 4), (5, 6)]

for a, b in pairs:
    print(f"a={a}, b={b}, sum={a+b}")
```

```python
# This pattern appears constantly with dict.items()
d = {"name": "Alice", "score": 95}

for key, value in d.items():
    print(f"{key}: {value}")
```

## Key Takeaways
