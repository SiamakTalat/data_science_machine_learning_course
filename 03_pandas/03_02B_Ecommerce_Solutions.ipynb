## Q10. American Express purchases over $95

```python
# Both conditions must be true — use & with parentheses around each
sum(
    (ecom['CC Provider'] == 'American Express') &
    (ecom['Purchase Price'] > 95)
)
```

---

## Q11. Cards expiring in 2025

```python
# Expiry stored as 'MM/YY' — extract the last two characters
sum(ecom['CC Exp Date'].apply(lambda x: x[3:] == '25'))
```

> **Explanation:** `x[3:]` slices from index 3 to end of the string `'MM/YY'`, returning the year part `'YY'`. We compare to `'25'` for 2025.

---

## Q12. Top 5 email providers

```python
# Split each email at '@', take the second part (domain), then count
ecom['Email'].apply(lambda x: x.split('@')[1]).value_counts().head(5)
```

> **Pattern:** `str.split('@')[1]` extracts the domain from any email string. This is a common text feature extraction technique.

---

## Key Takeaways

- `sum(boolean_series)` counts `True` values — clean idiom for conditional counting.
- `value_counts().head(n)` is the fastest way to find the top-N most frequent categories.
- `apply(lambda x: x[3:] == '25')` shows how string slicing inside a lambda enables efficient feature extraction.
- `apply(lambda x: x.split('@')[1])` + `value_counts()` extracts structured info from text fields — a foundational NLP-adjacent skill.
- Multi-condition filtering always requires `&` / `|` and parentheses around each condition.
