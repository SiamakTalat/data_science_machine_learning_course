## Q6. Highest paid person (full row)

```python
# idxmax() returns the index label of the maximum value
sal.loc[sal['TotalPayBenefits'].idxmax()]
```

---

## Q7. Lowest paid person (full row)

```python
sal.loc[sal['TotalPayBenefits'].idxmin()]
```

> **Observation:** The lowest TotalPayBenefits is likely a large negative number — not a genuine salary. This is a data quality issue common in public datasets (e.g., corrections, adjustments, or data entry errors).

## Q11. Job titles held by exactly one person in 2013

```python
# Step 1: filter to 2013 only
titles_2013 = sal[sal['Year'] == 2013]['JobTitle'].value_counts()

# Step 2: count how many titles have a count of exactly 1
sum(titles_2013 == 1)
```

---

## Q12. Employees with "chief" as a complete word in their title

```python
def chief_string(title):
    """
    Return True if 'chief' appears as a standalone word in the title.
    Lowercase + split avoids matching partial words like 'chieftain'.
    """
    return 'chief' in title.lower().split()

sum(sal['JobTitle'].apply(chief_string))
```

---

## Q13 (Bonus). Correlation between title length and TotalPayBenefits

```python
# Feature engineering: create a new column for title character count
sal['title_len'] = sal['JobTitle'].apply(len)

# Compute correlation matrix
sal[['title_len', 'TotalPayBenefits']].corr()
```

> **Interpretation:** A weak positive correlation would suggest longer titles tend to correspond to slightly higher pay — but the effect is small and likely not practically significant.

---

## Key Takeaways

- `idxmax()` and `idxmin()` return the **index label** of the extreme value — use with `.loc[]` to get the full row.
- `groupby('Year')['column'].mean()` is a clean one-liner for year-over-year aggregation.
- Always split strings before checking for word membership — `'chief' in title.split()` avoids partial matches.
- `apply()` with a custom function is the most flexible tool for string-based feature engineering.
- Data quality issues (negative salaries) appear in real-world datasets — always inspect anomalies critically.
- Feature engineering (title length) followed by `.corr()` is a foundational exploratory analysis pattern.
