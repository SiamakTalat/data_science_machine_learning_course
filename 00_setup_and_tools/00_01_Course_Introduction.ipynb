## 3. Jupyter Notebook — Productivity Shortcuts

Jupyter Notebook is the primary development environment throughout this bootcamp.
These shortcuts will save significant time during exercises and exploration.

````python
# Keyboard shortcuts to practice from day one:
#
#   Tab             → autocomplete method or variable names
#   Shift + Tab     → show docstring / inline help for current function
#   Shift + Enter   → run current cell and move to next
#   Alt + Enter     → run current cell and insert new cell below
#   Ctrl + S        → save notebook

# Try it: place your cursor inside len() and press Shift + Tab
len([1, 2, 3])
````

````python
# Useful terminal / notebook shell commands
# (run in a terminal or as a notebook magic command)

# pwd     → show current working directory
# cd      → change working directory
# jupyter notebook   → launch the notebook server

# Inside a notebook cell you can use the % magic prefix:
%pwd
````

**Expected behavior:** `%pwd` prints the directory where the notebook is stored. This is useful for loading datasets with relative paths.
