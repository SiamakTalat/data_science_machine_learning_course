## 2. Code Cells — The Core Unit of a Notebook

A **code cell** executes Python and displays the result immediately below it.

````python
# Code cells execute Python interactively
print("Hello from a code cell!")
1 + 1
````

**Two output behaviors:**
- `print()` explicitly displays output (works anywhere in the cell)
- The **last expression** in a cell is automatically displayed as the cell output

Only `1 + 1` appears as cell output above because it is the last expression. `print()` goes to stdout.

````python
# Intermediate expressions need print() to be visible
x = 10
print("x =", x)        # visible
y = x * 2
print("y =", y)        # visible
x + y                  # only this auto-displays as output
````

## 3. Keyboard Shortcuts

These are the most important shortcuts — learn them in the first session.

````python
# Shortcuts reference (not executable — use as a cheat sheet)
#
#  Shift + Enter   → run current cell, move to next cell
#  Alt + Enter     → run current cell, insert new cell below
#  Ctrl + S        → save notebook
#  Tab             → autocomplete
#  Shift + Tab     → show docstring for current function
#  Esc             → enter command mode (blue border)
#  Enter           → enter edit mode (green border)
#  A (command)     → insert cell above
#  B (command)     → insert cell below
#  D, D (command)  → delete current cell
#  M (command)     → change cell to markdown
#  Y (command)     → change cell to code

print("Shortcuts loaded — practice them during this session.")
````

> **Pro tip for exercises:** Use `Alt + Enter` to insert buffer cells so that your answer goes into a new cell and the expected output cell below it is not overwritten.

## 4. Cell Execution Order

Each code cell shows its execution number: `In [1]:`, `In [2]:`, etc.

This matters because:
- Variables defined in earlier cells are available to later cells
- Running cells **out of order** can cause confusing errors
- The number tells you the actual execution sequence, not the cell position

````python
# Cell executed first: defines a variable
message = "Execution order matters"
````

````python
# Cell executed second: uses the variable defined above
print(message)
````

> **Best practice:** Use **Kernel → Restart & Run All** regularly to verify your notebook runs cleanly from top to bottom.

## 5. Markdown Cells

**Markdown cells** render formatted text — headings, bullet points, bold, code snippets — not Python.

Switch a cell to markdown:
- Press `Esc` then `M` (command mode)
- Or use the dropdown menu in the toolbar

Common markdown syntax:

```markdown
# Heading 1
## Heading 2
**bold**
*italic*
- bullet point
`inline code`
```

---

## 6. Common Beginner Mistake — Wrong Cell Type

**Symptom:** Code has no syntax highlighting and does not execute — it just shows as plain text.

**Cause:** The cell was accidentally changed to a markdown cell.

**Fix:**
1. Click the cell
2. Press `Esc` then `Y` to convert it back to a code cell
3. Or use the toolbar dropdown and select **Code**

## 7. Saving and Exporting Notebooks

**Save:**
- `Ctrl + S` — manual save
- Jupyter autosaves periodically (look for the save indicator)
- **File → Save and Checkpoint** creates a restore point

**Export:**
- **File → Download As** supports:
  - `.py` — plain Python script
  - `.html` — shareable web page
  - `.pdf` — requires LaTeX
  - Other notebook formats

This makes notebooks reusable outside Jupyter — share HTML with non-technical stakeholders.

## 9. Virtual Environments with Conda (Optional)

A **virtual environment** is an isolated Python installation with its own packages.
This prevents version conflicts across projects.

**Why useful:**
- Project A needs NumPy 1.21, Project B needs NumPy 1.24 — no conflict
- Test new library versions without risking existing work
- Reproduce environments exactly on other machines

````python
# Conda environment commands (run in terminal, not inside a notebook)

# Create a new environment with NumPy
# conda create --name myenv numpy

# Activate the environment
# conda activate myenv       # macOS / Linux
# activate myenv             # Windows

# Install a package into the active environment
# conda install pandas

# List all environments
# conda info --envs

# Deactivate the environment
# conda deactivate

print("These commands run in your OS terminal — not inside a notebook.")
````

````python
# Create an environment with a specific Python version
# conda create --name py38env python=3.8 numpy pandas

# Practical demonstration of isolation:
# If numpy is installed in 'base' but NOT in 'myenv', then inside 'myenv':
#   import numpy   → ImportError
#   import pandas  → ImportError (unless installed there too)

print("Each environment is independent — packages must be installed per environment.")
````
