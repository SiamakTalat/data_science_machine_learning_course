## 1. Why VS Code for Data Science?

The browser-based Jupyter interface is functional but limited. VS Code wraps the same `.ipynb` notebook format inside a full professional editor, giving you tools that browser Jupyter cannot provide.

| Feature | Browser Jupyter | VS Code |
|---------|----------------|---------|
| Autocomplete | Basic | Full IntelliSense (type hints, docstrings) |
| Variable Explorer | No | Yes — inspect all live variables |
| DataFrame Viewer | No | Yes — spreadsheet-style view |
| Debugger | No | Yes — breakpoints inside cells |
| Git integration | No | Yes — built-in source control panel |
| Terminal | Separate window | Integrated panel |
| Refactoring tools | No | Yes — rename, extract, format |
| Multi-file editing | Tabs only | Full editor with split panes |
| Extensions | No | Hundreds available |

> **Key idea:** VS Code runs the same Jupyter kernel as browser Jupyter — the difference is entirely in the editing experience. Every `.ipynb` file works in both without any conversion.

## 4. Selecting the Python Interpreter and Kernel

The **kernel** (Python process) is selected per notebook. VS Code shows the current kernel in the top-right corner of the notebook editor.

**To change the kernel:**
1. Click the kernel name in the top-right corner of the notebook
2. A dropdown appears showing all available interpreters
3. Select the Conda environment or venv for this project

**To select the interpreter for `.py` files:**
- `Ctrl + Shift + P` → **Python: Select Interpreter**
- Choose from system Python, Conda envs, or venvs

```python
import sys

# Confirm which Python interpreter the current kernel is using
print("Interpreter:", sys.executable)
print("Version:    ", sys.version)
```

 > If `sys.executable` points to the wrong environment, click the kernel selector (top-right) and switch to the correct one before running any other cells.

## 5. Keyboard Shortcuts

VS Code uses its own shortcut layer on top of the standard Jupyter shortcuts. Most cell-execution shortcuts are identical; navigation and editing shortcuts differ.

```python
# VS Code notebook shortcuts (not executable — use as a cheat sheet)
#
#  CELL EXECUTION
#  ──────────────────────────────────────────────────────────
#  Shift + Enter        → run cell, move to next
#  Ctrl + Enter         → run cell, stay on same cell
#  Alt + Enter          → run cell, insert new cell below
#
#  CELL MANAGEMENT (click cell margin first to select)
#  ──────────────────────────────────────────────────────────
#  A                    → insert cell above
#  B                    → insert cell below
#  D, D                 → delete cell
#  M                    → change to markdown
#  Y                    → change to code
#  Z                    → undo cell action
#  Ctrl + Shift + P     → command palette (search all actions)
#
#  EDITOR (inside a cell)
#  ──────────────────────────────────────────────────────────
#  Ctrl + /             → comment / uncomment line
#  Tab                  → IntelliSense autocomplete
#  Shift + Tab          → show docstring for function under cursor
#  Ctrl + Space         → trigger autocomplete manually
#  Alt + ↑ / ↓          → move line up / down
#  Ctrl + D             → select next occurrence of word
#  F2                   → rename symbol across the file
#
#  NOTEBOOK-LEVEL
#  ──────────────────────────────────────────────────────────
#  Ctrl + Shift + P → "Run All Cells"       → run entire notebook
#  Ctrl + Shift + P → "Restart Kernel"      → reset kernel state
#  Ctrl + Shift + P → "Clear All Outputs"   → clean notebook for sharing

print("VS Code adds full editor power on top of standard Jupyter shortcuts.")
```

## 6. IntelliSense — Autocomplete and Inline Docs

IntelliSense is the biggest daily advantage of VS Code over browser Jupyter. It reads type information, docstrings, and imported libraries to give context-aware suggestions as you type.

```python
import pandas as pd

df = pd.DataFrame({
    'name':  ['Alice', 'Bob', 'Carol'],
    'score': [88, 92, 79],
    'grade': ['B', 'A', 'C']
})

# In VS Code: type  df.  and pause — IntelliSense shows every DataFrame method
# Hover over a method name to see its docstring inline
# Press Shift + Tab inside a function call to see the full signature

df.head()
```

**What IntelliSense provides that browser Jupyter does not:**
- Full method list with descriptions as you type
- Parameter hints inside function calls (names, types, defaults)
- Inline docstrings on hover — no need to look up documentation separately
- Import suggestions — type a module name and VS Code offers to add the import
- Error underlining before you run the cell

## 8. Debugging Notebook Cells

VS Code lets you set **breakpoints inside code cells** and step through execution line by line — impossible in browser Jupyter.

**How to debug a cell:**
1. Click in the left margin of a code line inside a cell to set a red breakpoint dot
2. Click the **Debug Cell** button that appears above the cell (or `Ctrl + Shift + P` → **Jupyter: Debug Current Cell**)
3. The debugger stops at the breakpoint — inspect variables in the Debug panel on the left
4. Use `F10` (step over), `F11` (step into), `F5` (continue) to navigate

```python
def compute_stats(values):
    total   = sum(values)       # set a breakpoint here in VS Code
    average = total / len(values)
    maximum = max(values)
    return {'total': total, 'average': average, 'maximum': maximum}

data = [15, 42, 8, 93, 27]
stats = compute_stats(data)
print(stats)
```

 > In browser Jupyter you can only add `print()` statements to understand what a function is doing. In VS Code you can pause inside the function and inspect every variable at that exact moment.

## 9. The Integrated Terminal

VS Code includes a full terminal panel — open with `Ctrl + `` ` (backtick). It starts in your project folder and uses the selected Python environment automatically.

**Common uses in a data science workflow:**

| Task | Command |
|------|---------|
| Install a package | `pip install seaborn` |
| Create a Conda env | `conda create --name myenv pandas numpy` |
| Activate Conda env | `conda activate myenv` |
| Run a Python script | `python my_script.py` |
| Start Jupyter server | `jupyter notebook` |
| Check installed packages | `pip list` |
| Git commit | `git add . && git commit -m "message"` |

> Having the terminal in the same window means you never need to context-switch to a separate shell. Install a package and continue in the notebook without restarting.

## 10. Git Integration

VS Code has a built-in **Source Control** panel (`Ctrl + Shift + G`) that wraps Git with a visual interface — no terminal Git commands required for everyday operations.

```python
# Git workflow in VS Code — what each panel does (reference only)
#
#  SOURCE CONTROL PANEL  (Ctrl + Shift + G)
#  ──────────────────────────────────────────────────────────
#  Changes section     → files modified since last commit
#  Stage button (+)    → equivalent to  git add <file>
#  Commit box          → type message + Ctrl+Enter = git commit
#  Sync button         → git pull then git push in one click
#
#  TIMELINE VIEW  (bottom of Explorer panel)
#  ──────────────────────────────────────────────────────────
#  Shows per-file commit history
#  Click any entry to diff that version against the current file
#
#  DIFF EDITOR
#  ──────────────────────────────────────────────────────────
#  Click any changed file in Source Control → side-by-side diff opens
#  Green = added lines, red = removed lines

print("Git is built into VS Code — no separate GitHub Desktop or terminal needed.")
```

## 11. Converting Between Notebooks and Python Scripts

VS Code makes it easy to move code between `.ipynb` notebooks and plain `.py` scripts — useful when a notebook analysis is ready to be turned into a reusable module.

```python
# Conversion options (run in the integrated terminal)
#
# Convert notebook → Python script
# jupyter nbconvert --to script my_notebook.ipynb
# → produces my_notebook.py with  # %%  cell markers
#
# Convert Python script → notebook (requires jupytext)
# pip install jupytext
# jupytext --to notebook my_script.py
#
# The # %% marker in a .py file is recognised by VS Code as a cell boundary
# — run cells in a .py file the same way as in a notebook

# Example .py file with cell markers:
example_script = """
# %%
import pandas as pd
df = pd.read_csv('data.csv')

# %%
df.describe()
"""
print("# %% in a .py file creates interactive cells in VS Code.")
```

## Key Takeaways

- VS Code runs the exact same Jupyter kernel as browser Jupyter — all `.ipynb` files work in both without conversion.
- The main advantages over browser Jupyter: **IntelliSense**, **Variable Explorer**, **Data Viewer**, **debugger**, and **integrated terminal**.
