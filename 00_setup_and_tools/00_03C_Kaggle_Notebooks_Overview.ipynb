## 1. What is Kaggle?

**Kaggle** is a platform owned by Google that combines three things in one place:
- A **dataset repository** — thousands of public datasets ready to use
- A **competition platform** — data science contests with prizes and leaderboards
- A **notebook environment** (Kaggle Notebooks, formerly Kernels) — cloud-hosted Jupyter notebooks with free GPU access

| Feature | Jupyter (local) | Google Colab | Kaggle Notebooks |
|---------|----------------|--------------|-----------------|
| Setup | Install locally | Browser only | Browser only |
| Storage | Local disk | Google Drive | Kaggle servers |
| GPU/TPU | Own hardware | Free (limited) | Free (30 hrs/week GPU) |
| Datasets | Manual download | Manual/Drive | One-click attach |
| Community | Private | Private | Public by default |
| Competition integration | No | No | Yes |
| Version history | Git | Manual | Built-in commits |

> **Key idea:** Kaggle Notebooks are Jupyter notebooks running on Kaggle's infrastructure, tightly integrated with Kaggle datasets and competitions.

## 2. Creating and Opening Notebooks

**To create a new notebook:**
1. Go to [kaggle.com](https://www.kaggle.com) and sign in (free account)
2. Click **Create** (top right) → **New Notebook**
3. A blank Jupyter-style notebook opens with a Python kernel

**To open an existing notebook:**
- Go to **Your profile → Code** to see your notebooks
- Browse **Competitions → [any competition] → Code** for public notebooks
- Click **Copy & Edit** (the fork button) on any public notebook to get your own editable copy

> Kaggle notebooks are **public by default** — be mindful of what you include before sharing.

## 4. Keyboard Shortcuts

Kaggle Notebooks share most shortcuts with Jupyter. The key difference is that Kaggle uses its own editor controls for some actions.

```python
# Kaggle Notebook keyboard shortcuts (not executable — use as a cheat sheet)
#
#  Shift + Enter        → run cell, move to next
#  Alt + Enter          → run cell, insert new cell below
#  Ctrl + Enter         → run cell in place (stay on same cell)
#  Ctrl + S             → save draft
#  Ctrl + /             → comment/uncomment selected lines
#  Tab                  → autocomplete
#  Shift + Tab          → show docstring
#  Ctrl + Shift + P     → open command palette
#  Esc → A              → insert cell above (command mode)
#  Esc → B              → insert cell below (command mode)
#  Esc → D, D           → delete cell (command mode)
#  Esc → M              → convert to markdown (command mode)
#  Esc → Y              → convert to code (command mode)

print("Shortcuts ready — standard Jupyter shortcuts work on Kaggle.")
```

## 5. Accessing Datasets Inside a Notebook

The biggest Kaggle advantage over Colab is **one-click dataset attachment**. No downloading, uploading, or Drive mounting required.

**To attach a dataset:**
1. In the right-hand panel, click **+ Add data**
2. Search for any Kaggle dataset or competition data
3. Click **Add** — the data appears at `/kaggle/input/<dataset-name>/`

All attached datasets are read-only and mounted instantly.

```python
import os

# List all attached datasets — only works inside a Kaggle notebook
input_path = '/kaggle/input'
if os.path.exists(input_path):
    for dataset in os.listdir(input_path):
        print(f"Dataset: {dataset}")
        dataset_path = os.path.join(input_path, dataset)
        for f in os.listdir(dataset_path):
            print(f"  └── {f}")
else:
    print("Not running on Kaggle — /kaggle/input does not exist here.")
```

## 6. The Kaggle API — Using Kaggle Data Locally

You can also download Kaggle datasets to your **local machine or Colab** using the Kaggle API.

**Setup (one-time):**
1. Go to [kaggle.com/settings](https://www.kaggle.com/settings) → **API** → **Create New Token**
2. A `kaggle.json` file downloads — it contains your username and API key
3. Place it at `~/.kaggle/kaggle.json` (macOS/Linux) or `C:\Users\<you>\.kaggle\kaggle.json` (Windows)

## 7. GPU and TPU Runtimes

Kaggle provides **30 hours of free GPU per week** — more generous than Colab's free tier for sustained workloads.

**To enable GPU:**
1. In the right-hand **Settings** panel, find **Accelerator**
2. Select **GPU T4 x2** or **GPU P100**
3. The session restarts with GPU hardware attached

**To enable TPU:**
- Select **TPU VM v3-8** from the same dropdown

> GPU hours reset weekly. Check your remaining hours at [kaggle.com/settings](https://www.kaggle.com/settings).

```python
import subprocess

# Check whether a GPU is available in this session
result = subprocess.run(['nvidia-smi'], capture_output=True, text=True)
if result.returncode == 0:
    print("GPU detected:")
    print(result.stdout[:300])
else:
    print("No GPU detected — accelerator is set to None.")
    print("Enable GPU in the Settings panel on the right.")
```

## 8. Saving, Committing, and Version Control

Kaggle has a built-in versioning system that works differently from Jupyter or Colab.

| Action | What it does |
|--------|-------------|
| **Save draft** (`Ctrl + S`) | Saves current state — no public version created |
| **Save Version** (blue button) | Creates a numbered, permanent snapshot |
| **Quick Save** | Commits with the current output (fast) |
| **Save & Run All** | Restarts the kernel and runs the full notebook before committing |

> **Best practice:** Always use **Save & Run All** before submitting to a competition — it guarantees the committed output is reproducible.

```python
# Output files go to /kaggle/working/ — this is what gets saved on commit
output_path = '/kaggle/working'
if os.path.exists(output_path):
    print("Output directory:", output_path)
    print("Files:", os.listdir(output_path))
else:
    print("Not on Kaggle — /kaggle/working does not exist here.")
    print("On Kaggle, write output files here:")
    print("  df.to_csv('/kaggle/working/submission.csv', index=False)")
```

## 10. Forking and Sharing Public Notebooks

Kaggle has a strong community of public notebooks — learning from others is a core part of the platform.

**Forking (Copy & Edit):**
- Open any public notebook and click **Copy & Edit** to get your own editable copy
- Your fork is private by default — you control when or whether it becomes public
- Forked notebooks track their origin — you can compare versions

**Making your notebook public:**
1. Click the notebook **title bar menu** → **Settings**
2. Toggle **Privacy** from Private to Public
3. Others can then find it, upvote it, and fork it

> Upvoted public notebooks appear in community rankings — useful for building a portfolio.
