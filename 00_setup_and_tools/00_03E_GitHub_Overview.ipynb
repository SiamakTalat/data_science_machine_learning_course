## 1. What is GitHub?

**GitHub** is a web platform for hosting code repositories. It is built on top of Git (the version control system), but most of what makes GitHub valuable is its web interface and community features — not Git itself.

| Layer | What it is |
|-------|-----------|
| **Git** | A version control tool — tracks changes to files locally |
| **GitHub** | A web platform — hosts repositories, enables collaboration, and adds social features |

Think of Git as the engine and GitHub as the car — you can drive the car (GitHub) without understanding the full engine (Git commands).

**Why data scientists use GitHub:**
- Share notebooks and analysis with reproducible code
- Find and reuse open-source ML projects and datasets
- Follow researchers and track new work in the field
- Build a public portfolio of projects
- Collaborate with teammates on the same codebase
- Connect with the VS Code, Colab, and Kaggle ecosystems

## 3. Navigating a Repository

A **repository** (repo) is a folder of files hosted on GitHub, with full version history attached. Understanding how to read a repo is a core skill for finding and reusing open-source work.

**Anatomy of a GitHub repository page:**

| Element | Where | What it shows |
|---------|-------|--------------|
| File browser | Centre | All files and folders in the project |
| `README.md` | Below files | Auto-rendered documentation |
| **Code** tab | Top nav | File browser (default view) |
| **Issues** tab | Top nav | Bug reports and feature requests |
| **Pull requests** | Top nav | Proposed changes under review |
| **Actions** tab | Top nav | Automated CI/CD workflows |
| **Releases** | Right sidebar | Tagged stable versions |
| **Stars** count | Top right | Community popularity signal |
| **Fork** button | Top right | Copy to your own account |
| Commit count | Above files | Total number of saved changes |
| Branch selector | Above files | Switch between versions of the code |

````python
# GitHub URL patterns — understanding the structure helps you navigate quickly
#
# Repository home
#   github.com/<user>/<repo>
#
# A specific file
#   github.com/<user>/<repo>/blob/<branch>/<path/to/file>
#
# Raw file content (downloadable, usable in pd.read_csv)
#   raw.githubusercontent.com/<user>/<repo>/<branch>/<path/to/file>
#
# Commit history for the whole repo
#   github.com/<user>/<repo>/commits
#
# Commit history for a single file
#   github.com/<user>/<repo>/commits/<branch>/<path/to/file>
#
# Issues list
#   github.com/<user>/<repo>/issues
#
# Releases
#   github.com/<user>/<repo>/releases

print("Knowing the URL structure lets you navigate directly without clicking through the UI.")
````

## 4. Reading Data Directly from GitHub

CSV and other data files hosted on GitHub can be loaded into Pandas using the **raw** URL — no downloading required.

````python
import pandas as pd

# On a GitHub file page: click the "Raw" button to get the raw URL
# Change  github.com  →  raw.githubusercontent.com
# Remove  /blob/  from the path

# Example: a public CSV hosted on GitHub
url = 'https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv'
df  = pd.read_csv(url)

print("Shape:", df.shape)
df.head()
````

 > This works in any environment — local Jupyter, VS Code, Colab, or Kaggle — as long as internet access is available. It is the fastest way to share a dataset alongside a notebook without bundling the file.

## 6. Forking a Repository

**Forking** creates a complete copy of someone else's repository under your own GitHub account. Your fork is independent — changes you make do not affect the original.

**When to fork:**
- You want to experiment with an existing project without affecting the original
- You want to customise a template or starter project for your own use
- You intend to contribute back (you fork, make changes, then open a Pull Request)

**How to fork:**
1. Go to any public repository on GitHub
2. Click the **Fork** button (top right)
3. Choose your account as the destination
4. Your copy appears at `github.com/your-username/repo-name`

> Forking differs from cloning. A **fork** lives on GitHub (remote copy). A **clone** downloads a repo to your local machine. Both are covered further in the Git section.

## 7. GitHub and Jupyter Notebooks

GitHub automatically **renders `.ipynb` files** as formatted pages — outputs, plots, and markdown all display correctly in the browser without any special setup.

**Practical implications:**
- Share a notebook by pasting its GitHub URL — anyone can read it without running anything
- Plots and tables are embedded in the rendered view
- The raw JSON of the notebook is still accessible via the **Raw** button

**Limitation:** The rendered view is static — code cannot be run. For interactive notebooks, use the integrations below.

````python
# Opening a GitHub notebook in other environments
#
# IN GOOGLE COLAB
# ────────────────────────────────────────────────────────────────
# Method 1 — From the Colab open dialog:
#   File → Open notebook → GitHub tab → paste the repo URL → select the .ipynb file
#
# Method 2 — URL trick (fastest):
#   Replace  github.com  with  colab.research.google.com/github  in the notebook URL
#   github.com/user/repo/blob/main/notebook.ipynb
#   →  colab.research.google.com/github/user/repo/blob/main/notebook.ipynb
#
# IN KAGGLE
# ────────────────────────────────────────────────────────────────
#   Create notebook → File → Import notebook → paste the GitHub URL
#
# IN VS CODE
# ────────────────────────────────────────────────────────────────
#   Clone the repo locally (Git section covers this), then open the .ipynb file

print("The Colab URL trick is the fastest way to make any public GitHub notebook runnable.")
````

## 8. GitHub Gists — Sharing Code Snippets

**Gists** are lightweight single-file repositories — ideal for sharing a helper function, a quick analysis, or a configuration file without creating a full repository.

**To create a Gist:**
1. Go to [gist.github.com](https://gist.github.com)
2. Paste your code, give it a description and filename (include the extension: `.py`, `.ipynb`, `.csv`)
3. Choose **Public** (discoverable) or **Secret** (only visible via direct link)
4. Click **Create public gist** or **Create secret gist**

**What makes Gists useful:**
- Instant shareable link — paste in Slack, email, or Stack Overflow
- Syntax highlighting for any language
- `.ipynb` Gists are rendered as formatted notebooks
- Supports multiple files in one Gist
- Can be embedded in web pages with a `<script>` tag

````python
# Example: code worth sharing as a Gist
# A reusable function — save this as a .py Gist to reference from other projects

def summarise_dataframe(df):
    """Return a summary table of column types, nulls, and unique counts."""
    import pandas as pd
    summary = pd.DataFrame({
        'dtype':   df.dtypes,
        'nulls':   df.isna().sum(),
        'null_%':  (df.isna().mean() * 100).round(1),
        'unique':  df.nunique()
    })
    return summary

import pandas as pd
demo = pd.DataFrame({'a': [1, 2, None], 'b': ['x', 'x', 'y'], 'c': [1.1, 2.2, 3.3]})
summarise_dataframe(demo)
````

## 9. Writing an Effective README

The `README.md` file is the front page of your repository — it renders automatically on the repo home page. A clear README is the difference between a project someone explores and one they skip.

**Structure for a data science project README:**

```markdown
# Project Title

One sentence describing what this project does and why.

## Problem Statement
What question does this project answer?

## Dataset
- Source: [Kaggle / UCI / etc.]
- Size: rows × columns
- Key features: list important columns

## Approach
- EDA → Feature Engineering → Modelling → Evaluation

## Results
- Key finding or model metric (e.g., Test accuracy: 92%)
- Include a plot or table if possible

## How to Run
pip install -r requirements.txt
jupyter notebook notebooks/analysis.ipynb

## File Structure
├── data/           raw and processed data
├── notebooks/      Jupyter notebooks
├── src/            reusable Python modules
└── README.md
```
