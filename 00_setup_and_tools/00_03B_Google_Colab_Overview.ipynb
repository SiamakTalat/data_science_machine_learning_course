## 1. What is Google Colab?

**Google Colaboratory (Colab)** is a free, cloud-hosted Jupyter notebook environment provided by Google. You run it entirely in the browser — no installation required.

| Feature | Jupyter (local) | Google Colab |
|---------|----------------|--------------|
| Setup | Install Python + Jupyter | Open a browser |
| Storage | Local disk | Google Drive |
| Compute | Your CPU/RAM | Google's servers |
| GPU/TPU | Only if you have one | Free tier available |
| Sharing | Export file manually | Share link instantly |
| Internet access | Full | Full |

> **Key idea:** Colab notebooks **are** Jupyter notebooks (`.ipynb` format) — the cell structure, shortcuts, and kernel concept are identical. What differs is where they run and how they store data.

## 2. Opening and Creating Notebooks

**To start a new notebook:**
1. Go to [colab.research.google.com](https://colab.research.google.com)
2. Sign in with a Google account
3. Click **New notebook** — or **File → New notebook** from within Colab

**To open an existing notebook:**
- **From Google Drive:** `File → Open notebook → Google Drive`
- **From GitHub:** `File → Open notebook → GitHub` — paste a repo URL
- **Upload a local `.ipynb`:** `File → Open notebook → Upload`

> Notebooks saved in Colab are stored in **My Drive → Colab Notebooks** by default.

## 3. Code Cells and Markdown Cells

Colab uses the same two cell types as Jupyter.

```python
# Code cells work exactly as in Jupyter
print("Running on Google Colab!")
2 ** 10
```

**Cell type rules (same as Jupyter):**

- Add cells using the **+ Code** and **+ Text** buttons at the top, or hover between cells

## 6. Installing Packages

Colab comes with most data science libraries pre-installed (NumPy, Pandas, Matplotlib, Scikit-learn, TensorFlow, etc.).

For packages that are missing, install them with `pip` using a `!` shell prefix.

```python
# The ! prefix runs a shell command inside a code cell
# !pip install some-package

# Check what version of a package is installed
import numpy
print("NumPy version:", numpy.__version__)
```

> **Important:** Installed packages are lost when the runtime resets. Place `!pip install` cells near the top of the notebook so they re-run automatically each session.

## 7. Using GPU and TPU Runtimes

Colab offers free access to GPU and TPU accelerators — critical for deep learning training.

**To enable a GPU:**
1. **Runtime → Change runtime type**
2. Under *Hardware accelerator*, select **GPU** (or **TPU**)
3. Click **Save** — the runtime restarts with the new hardware

> The free tier has usage limits. Sessions disconnect after ~90 minutes of inactivity or ~12 hours total.

```python
import subprocess

# Check whether a GPU is available in this runtime
result = subprocess.run(['nvidia-smi'], capture_output=True, text=True)
if result.returncode == 0:
    print("GPU detected:")
    print(result.stdout[:300])
else:
    print("No GPU detected — runtime is CPU only.")
    print("Go to Runtime → Change runtime type to enable a GPU.")
```

## 9. Runtime Management

The Colab **runtime** is equivalent to the Jupyter **kernel** — the live Python process behind the notebook.

```python
# Runtime management reference (use the Runtime menu, not code)
#
#  Runtime → Run all              → runs all cells top to bottom
#  Runtime → Interrupt execution  → stops a running cell (like Kernel → Interrupt)
#  Runtime → Restart runtime      → clears all variables and state
#  Runtime → Restart and run all  → clean run from top to bottom
#  Runtime → Disconnect and delete runtime → free the VM immediately

print("Use Runtime → Restart and run all to verify the notebook runs cleanly.")
```

**Runtime behaviour to know:**

| Situation | Action |
|-----------|--------|
| Cell runs forever | Runtime → Interrupt execution |
| Variables in a bad state | Runtime → Restart runtime |
| Want a verified fresh run | Runtime → Restart and run all |
| Switch to GPU/TPU | Runtime → Change runtime type |
| Session idle too long | Colab disconnects automatically (~90 min) |

## 10. Sharing and Exporting Notebooks

**Sharing:**
- Click the **Share** button (top right) — same as sharing a Google Doc
- Set permissions: *Viewer*, *Commenter*, or *Editor*
- Anyone with the link can view (or edit, depending on settings)

**Exporting:**
- **File → Download → Download .ipynb** — standard notebook format, compatible with Jupyter
- **File → Download → Download .py** — plain Python script
- **File → Print** — saves as PDF via the browser print dialog

> Colab notebooks are stored on Google Drive as `.ipynb` files, so they can be opened directly in local Jupyter at any time.

## 11. Colab vs Jupyter — Practical Differences

Understanding the trade-offs helps you choose the right tool for each situation.

```python
# Summary of practical differences (reference only)
#
#  COLAB ADVANTAGES
#  ─────────────────────────────────────────────────────────
#  No installation required        → works anywhere with a browser
#  Free GPU/TPU access             → essential for deep learning
#  Instant sharing                 → like Google Docs
#  Pre-installed libraries         → saves environment setup time
#  Runs in the cloud               → no load on your local machine
#
#  COLAB LIMITATIONS
#  ─────────────────────────────────────────────────────────
#  Session resets after ~12 hours  → installed packages and /content/ lost
#  Idle disconnect after ~90 min   → long jobs must be managed carefully
#  Slower file access without Drive mount
#  Requires internet               → no offline use
#  Free RAM is limited             → very large datasets may crash

print("Choose Colab for GPU tasks, collaboration, and no-install access.")
```

## Key Takeaways

- Google Colab is a **cloud-hosted Jupyter environment** — same `.ipynb` format, same cell types, same shortcuts.
- No installation required — open a browser, sign in, and start coding.

- **Mount Google Drive** with `drive.mount('/content/drive')` to access files persistently.
- Use **`!pip install`** for missing packages — place these cells at the top of the notebook.
