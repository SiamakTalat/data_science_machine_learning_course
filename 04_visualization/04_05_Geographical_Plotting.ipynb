## Part 1 — Introduction to Geographical Plotting

Geographical plotting is often more difficult than ordinary plotting because the data can come in many different formats:
- Country names or codes
- State abbreviations or FIPS codes
- County identifiers
- Latitude / longitude pairs
- Shapefiles or GeoJSON

Because of this complexity, geographical plotting usually requires careful preprocessing, reference to documentation, and exact syntax for the chosen library.

In this course we focus on **Plotly geographical plotting** because:
- It produces interactive maps
- It integrates well with notebook workflows
- It allows hover information
- It provides visually impressive results with relatively little code

> Matplotlib also has a `basemap` extension for static maps, but we emphasise Plotly here for its interactivity.

## Part 3 — Basic USA Choropleth Example

A **choropleth map** colors geographical regions according to numerical values. We start with a simple USA example using state abbreviations.

```python
data = dict(
    type='choropleth',
    locations=['AZ', 'CA', 'NY'],
    locationmode='USA-states',
    colorscale='Portland',
    text=['Arizona', 'California', 'New York'],
    z=[1.0, 2.0, 3.0],
    colorbar={'title': 'Example Values'}
)
```

```python
layout = dict(
    geo={'scope': 'usa'}
)
```

```python
choromap = go.Figure(data=[data], layout=layout)
iplot(choromap)
```

**Key pieces of the `data` dictionary:**

| Key | Role |
|-----|------|
| `type` | Must be `'choropleth'` for this map type |
| `locations` | Regional identifiers (state codes, country codes, etc.) |
| `locationmode` | How Plotly interprets the identifiers (`'USA-states'`, `'country names'`, `'ISO-3'`) |
| `text` | Hover text |
| `z` | Numeric values that determine the colour of each region |
| `colorscale` | Colour gradient |
| `colorbar` | Legend configuration |

## Part 5 — USA Choropleth with Real Data

Now we work with a real dataset: **agricultural exports by state** (2011).

```python
df = pd.read_csv('2011_US_AGRI_Exports')
df.head()
```

Important columns:
- `code` — state abbreviation
- `state` — state name
- `total exports` — numeric variable to visualise
- `text` — hover text

```python
data = dict(
    type='choropleth',
    colorscale='YlOrRd',
    locations=df['code'],
    locationmode='USA-states',
    z=df['total exports'],
    text=df['text'],
    colorbar={'title': 'Millions USD'},
    marker=dict(
        line=dict(
            color='rgb(255,255,255)',
            width=1
        )
    )
)
```

```python
layout = dict(
    title='2011 US Agriculture Exports by State',
    geo=dict(
        scope='usa',
        showlakes=True,
        lakecolor='rgb(85,173,240)'
    )
)
```

```python
choromap = go.Figure(data=[data], layout=layout)
iplot(choromap)
```

**Additional features used here:**
- `marker.line.color` — controls state border colour
- `marker.line.width` — controls border thickness
- `showlakes=True` — adds lakes to the map
- `lakecolor` — customises lake colour
- `title` — labels the full map

---

## Part 6 — Border Customisation

Map borders can be emphasised or softened depending on the audience and purpose. Thicker borders can help regional separation, but too much emphasis makes the map visually heavy.

```python
data = dict(
    type='choropleth',
    colorscale='YlOrRd',
    locations=df['code'],
    locationmode='USA-states',
    z=df['total exports'],
    text=df['text'],
    colorbar={'title': 'Millions USD'},
    marker=dict(
        line=dict(
            color='rgb(0,0,0)',
            width=2
        )
    )
)
```

```python
choromap = go.Figure(data=[data], layout=layout)
iplot(choromap)
```

## Part 7 — World Choropleth Example

Now we move from national scale to international scale using world GDP data.

```python
df = pd.read_csv('2014_World_GDP')
df.head()
```

Key columns:
- `COUNTRY` — country name
- `GDP (BILLIONS)` — numeric variable
- `CODE` — ISO-3 country code

```python
data = dict(
    type='choropleth',
    locations=df['CODE'],
    z=df['GDP (BILLIONS)'],
    text=df['COUNTRY'],
    colorbar={'title': 'GDP in Billions USD'}
)
```

```python
layout = dict(
    title='2014 Global GDP',
    geo=dict(
        showframe=False,
        projection={'type': 'mercator'}
    )
)
```

```python
choromap = go.Figure(data=[data], layout=layout)
iplot(choromap)
```

> For world maps, `locations` uses ISO-3 country codes and no `locationmode` is needed (it defaults to `'ISO-3'`).

---

## Part 8 — World Map Projection Styles

Plotly supports several projection styles. Different projections are useful for different communication goals:

| Projection | Character |
|------------|----------|
| `mercator` | Familiar, web-standard (distorts area at poles) |
| `natural earth` | Balanced, good for general presentations |
| `orthographic` | Globe-like, good for hemisphere focus |
| `robinson` | Compromise between area and shape distortion |
| `miller` | Similar to Mercator, less polar distortion |

```python
layout = dict(
    title='2014 Global GDP — Natural Earth',
    geo=dict(
        showframe=False,
        projection={'type': 'natural earth'}
    )
)
```

```python
choromap = go.Figure(data=[data], layout=layout)
iplot(choromap)
```

## Part 9 — Why Documentation Matters Here

Geographical plotting syntax is harder to memorise than ordinary plotting syntax. In practice, real workflows usually require:
1. Starting from a known example
2. Adapting the `data` dictionary
3. Adapting the `layout` dictionary
4. Checking the output
5. Refining colours, borders, and hover text

> It is completely normal to work by reference — keep a working example handy and modify from there.

## Part 10 — Practical Interpretation Notes

When building choropleth maps, always consider:
- **What exactly is encoded by colour** — is it clear to the viewer?
- **Whether darker means larger or smaller** — some colorscales reverse expectations
- **Whether the colorscale is intuitive** — sequential for magnitudes, diverging for positive/negative
- **Whether the geographical unit makes sense** — state vs county vs country
- **Whether interactivity adds meaningful value** — hover text is powerful but needs useful content

**Good use cases for choropleth maps:**
- GDP by country
- Exports by state
- Election data by region
- Population density
- Energy consumption
- Public health indicators

## Key Takeaways

- Geographical plotting is harder because **data formats vary widely** (names, codes, coordinates, shapefiles).
- Plotly provides **interactive choropleth maps** with zoom, hover, and pan.
- Choropleth construction relies on correctly structured `data` and `layout` dictionaries.
- The core `data` keys are: `type`, `locations`, `locationmode`, `text`, `z`, `colorscale`, `colorbar`.
- Map customisation includes **colorscales**, **borders** (`marker.line`), **lakes**, and **projections**.
- For USA maps: `locationmode='USA-states'` + `geo.scope='usa'`.
- For world maps: ISO-3 codes + a `projection` type.
- **Documentation is essential** — geographical plotting is always a reference-driven workflow.
