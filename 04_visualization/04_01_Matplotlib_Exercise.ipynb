## Reflection

Key ideas reinforced by these exercises:

- The **figure → axes → plot** hierarchy is the core mental model for Matplotlib.
- `add_axes([l, b, w, h])` gives precise positional control for inset layouts.
- `plt.subplots(nrows, ncols)` is the cleaner approach for regular grids.
- `axes[i]` accesses individual panels — treat it as a list of axes objects.
