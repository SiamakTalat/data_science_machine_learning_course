## Setup

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

x = np.arange(0, 100)
y = x * 2
z = x ** 2
```

---

## Exercise 1 — Basic Figure with Labels

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, y)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title')
```

---

## Exercise 2 — Figure with Two Axes (Main + Inset)

```python
fig = plt.figure()

# Large main plot: full canvas
ax1 = fig.add_axes([0, 0, 1, 1])

# Inset: positioned inside the main plot area
ax2 = fig.add_axes([0.2, 0.5, 0.2, 0.2])

ax1.plot(x, y)
ax2.plot(x, y)
```

---

## Exercise 3 — Inset with Axis Limits and Labels

```python
fig = plt.figure()

ax1 = fig.add_axes([0, 0, 1, 1])      # main plot: x vs z
ax2 = fig.add_axes([0.2, 0.5, 0.4, 0.4])  # inset: x vs y, zoomed

ax1.plot(x, z)
ax1.set_xlabel('X')
ax1.set_ylabel('Z')

ax2.plot(x, y)
ax2.set_title('zoom')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_xlim([20, 22])
ax2.set_ylim([30, 50])
```

> **Note on Exercise 3:** `set_xlim([20, 22])` zooms the inset into x-values 20–22, while the outer plot shows the full range. This is the classic "magnified detail" layout pattern.

---

## Exercise 4 — Subplots with Styling

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 2))

axes[0].plot(x, y, color='blue', lw=3, ls='--')
axes[1].plot(x, z, color='red',  lw=3)

plt.tight_layout()
```

---

## Key Takeaways

- `fig.add_axes([l, b, w, h])` gives precise control: values are fractions of the canvas.
- `plt.subplots(1, 2)` is the cleaner pattern for regular side-by-side panels.
- `figsize=(12, 2)` stretches the figure wide — adjust to suit data shape.
- `set_xlim()` / `set_ylim()` on an inset creates a zoom window.
- Always use `plt.tight_layout()` with multiple subplots.
