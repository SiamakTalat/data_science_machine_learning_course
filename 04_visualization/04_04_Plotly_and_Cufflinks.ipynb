## Part 1 — Introduction

Plotly is an **interactive visualization library**. Unlike Matplotlib and Seaborn, Plotly plots are:
- Zoomable
- Pannable
- Hover-enabled (see exact values on mouseover)
- Clickable legends (toggle series on/off)
- Embeddable in dashboards

**Cufflinks** acts as a bridge between Pandas and Plotly, allowing interactive plotting using syntax very similar to Pandas `.plot()`:

| Static (Pandas) | Interactive (Cufflinks) |
|-----------------|------------------------|
| `df.plot()` | `df.iplot()` |
| `df.plot.hist()` | `df.iplot(kind='hist')` |
| `df.plot.scatter(x, y)` | `df.iplot(kind='scatter', x, y)` |

This makes interactive data exploration extremely fast.

## Part 5 — Interactive Line Plot

The simplest interactive plot — hover over any point to see exact values, click legend entries to toggle columns.

```python
df.iplot()
```

**Key interactive features visible in this plot:**
- Zoom (click-drag to select a region)
- Hover values (move cursor over points)
- Toggle columns on/off (click legend entries)
- Save figure (camera icon in toolbar)
- Pan and reset axes (toolbar buttons)

---

## Part 6 — Interactive Scatter Plot

For scatter plots, `mode='markers'` is required to avoid connecting points with lines.

```python
df.iplot(
    kind='scatter',
    x='A',
    y='B',
    mode='markers'
)
```

---

## Part 7 — Scatter Plot with Larger Markers

```python
df.iplot(
    kind='scatter',
    x='A',
    y='B',
    mode='markers',
    size=10
)
```

---

## Part 8 — Interactive Bar Plot

Ideal for interactive business dashboards and category summaries.

```python
df2.iplot(
    kind='bar',
    x='Category',
    y='Values'
)
```

---

## Part 9 — Aggregated Bar Plot

Often bar plots are more useful after aggregation — a very common quick-summary workflow.

```python
df.sum().iplot(kind='bar')
```

---

## Part 10 — Interactive Box Plot

Useful for spread comparison, outlier detection, feature inspection, and comparing multiple columns interactively.

```python
df.iplot(kind='box')
```

## Part 11 — 3D Surface Plot

Plotly supports interactive 3D surfaces — you can click and drag to rotate the view.

```python
df3 = pd.DataFrame({
    'x': [1, 2, 3, 4, 5],
    'y': [10, 20, 30, 20, 10],
    'z': [5, 4, 3, 2, 1]
})
```

```python
df3
```

```python
df3.iplot(kind='surface')
```

---

## Part 12 — Surface Plot with Colorscale

Plotly colorscales differ from Matplotlib colormaps. Common choices: `'rdylbu'`, `'blues'`, `'greens'`, `'reds'`.

```python
df3.iplot(
    kind='surface',
    colorscale='rdylbu'
)
```

## Part 16 — Bubble Plot

## Part 18 — When to Use Plotly

| Scenario | Best tool |
|----------|----------|
| Quick static EDA | Pandas `.plot` |
| Statistical plots, grouped comparisons | Seaborn |
| Fine-grained layout control, publication figures | Matplotlib |
| Interactive exploration, dashboards, presentations | **Plotly / Cufflinks** |

Plotly is especially useful when:
- Presenting to stakeholders who want to explore the data themselves
- Building dashboards (Plotly Dash)
- Inspecting dense data interactively
- Financial analysis (spread, candlestick, moving averages)
- Creating notebook demos for teaching

## Key Takeaways

- **Plotly** provides interactive plotting — zoom, hover, pan, toggle.
- **Cufflinks** connects Plotly directly to Pandas — use `df.iplot()` instead of `df.plot()`.
- Interactive plots are ideal for **exploration and dashboards**.
- Scatter, bar, histogram, box, spread, bubble, and 3D surface plots are especially powerful.
- Hover + zoom **dramatically improve data understanding** compared to static figures.
- Use `mode='markers'` for scatter plots to avoid connected lines.
- Plotly is the go-to for **presentations** and **stakeholder-facing** work.
