---

## Key Takeaways

- Choropleth maps require correctly structured `data` and `layout` dictionaries.
- The `locationmode` must match the format of your location data (country names vs state abbreviations vs ISO codes).
- World maps use `locationmode='country names'` or ISO-3 codes; USA maps use `locationmode='USA-states'`.
- Hover text, color bars, and projections all improve map readability.
- Working from reference examples is a normal and efficient workflow for geographical plotting.
