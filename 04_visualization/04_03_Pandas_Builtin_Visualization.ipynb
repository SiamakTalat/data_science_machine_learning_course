## Part 1 — Introduction

Pandas includes built-in visualization methods that sit on top of Matplotlib. These methods make it easy to create quick exploratory plots directly from a DataFrame or Series.

This is especially useful for:
- Quick EDA (Exploratory Data Analysis)
- Checking distributions
- Feature inspection
- Simple time series visualization
- Debugging preprocessing steps

> Pandas plotting is not usually the best choice for polished presentation figures, but it is extremely useful for **speed and convenience** during analysis.

---

## Part 2 — Setup

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

---

## Part 3 — Load Example Data

For this section we use a few CSV files stored alongside the notebook.

```python
df1 = pd.read_csv('df1', index_col=0)
df2 = pd.read_csv('df2')
```

```python
df1.head()
```

```python
df2.head()
```

Notice:
- `df1` has an index column already set
- `df2` uses a simple default integer index

---

## Part 4 — Quick Histogram

A fast histogram can be created directly from a Series — one of the most useful built-in Pandas visualizations for fast feature inspection.

```python
df1['A'].hist()
```

---

## Part 5 — Histogram with More Bins

Changing the number of bins gives more control over the displayed distribution.

```python
df1['A'].hist(bins=30)
```

---

## Part 6 — The General `.plot()` Interface

Pandas also supports a general `plot()` interface with a `kind` argument. All three calls below produce the same output:

```python
# Using the kind argument
df1['A'].plot(kind='hist', bins=30)
```

```python
# Using the accessor style (preferred — more readable)
df1['A'].plot.hist(bins=30)
```

> **Style tip:** the `.plot.hist()` accessor form is cleaner and more readable than passing `kind=`.

---

## Part 7 — Area Plot

Area plots fill the region under the lines — useful for comparing trends where filled area improves readability.

```python
df2.plot.area(alpha=0.4)
```

---

## Part 8 — Bar Plot

Bar plots treat the DataFrame index as the category axis.

```python
df2.plot.bar()
```

```python
# Stacked version — useful for comparing cumulative category contributions
df2.plot.bar(stacked=True)
```

---

## Part 9 — Line Plot

Line plots are especially useful for ordered data and time series.

```python
df1.plot.line(y='B', figsize=(12, 3), lw=1)
```

---

## Part 10 — Scatter Plot

Scatter plots help visualize relationships between two numerical variables.

```python
df1.plot.scatter(x='A', y='B')
```

---

## Part 11 — Scatter Plot with Color Dimension

A third variable can be encoded using color — a lightweight way to show three dimensions in a 2D plot.

```python
df1.plot.scatter(x='A', y='B', c='C', cmap='coolwarm')
```

---

## Part 12 — Scatter Plot with Size Dimension

A third variable can also be represented by point size.

```python
df1.plot.scatter(x='A', y='B', s=df1['C'] * 100)
```

---

## Part 13 — Box Plot

Box plots are useful for:
- Comparing spread across columns
- Identifying outliers
- Inspecting skewness
- Checking feature distributions quickly

```python
df2.plot.box()
```

---

## Part 14 — Hexbin Plot

For dense scatter data, hexbin plots often work better than ordinary scatter plots. The hexagons become darker where more points fall into the same region.

```python
df = pd.DataFrame(np.random.randn(1000, 2), columns=['A', 'B'])
```

```python
df.plot.hexbin(x='A', y='B', gridsize=25, cmap='coolwarm')
```

---

## Part 15 — KDE Plot

Kernel Density Estimation provides a smooth estimate of distribution shape.

```python
df1['A'].plot.kde()
```

```python
# Equivalent call
df1['A'].plot.density()
```

> `.kde()` and `.density()` are identical — use whichever reads better in context.

---

## Part 16 — KDE for Multiple Columns

Calling `.plot.kde()` on the full DataFrame overlays smooth density estimates for every column.

```python
df1.plot.kde()
```

---

## Part 17 — Matplotlib Stylesheets

Pandas plotting respects Matplotlib stylesheets — applying one often gives a cleaner look immediately.

```python
plt.style.use('ggplot')
df1['A'].hist()
```

---

## Part 18 — Popular Style Examples

Try these with your own plots to see which style best matches the context.

```python
plt.style.use('ggplot')
df1['A'].plot.hist(bins=30)
plt.title('ggplot')
```

```python
plt.style.use('bmh')
df1['A'].plot.hist(bins=30)
plt.title('bmh')
```

```python
plt.style.use('dark_background')
df1['A'].plot.hist(bins=30)
plt.title('dark_background')
```

```python
plt.style.use('fivethirtyeight')
df1['A'].plot.hist(bins=30)
plt.title('fivethirtyeight')
```

```python
# Reset to default for the rest of the notebook
plt.style.use('default')
```

## Part 19 — Area Plot on a Slice of Data

Sometimes we want to limit plotting to only part of a DataFrame — useful for the first part of a time sequence or a manageable subset.

```python
df3 = pd.read_csv('df3')
df3.head()
```

```python
df3.iloc[:30].plot.area(alpha=0.4)
```

---

## Part 20 — Legend Outside the Plot

Sometimes legends overlap the figure. We can move them outside — especially helpful when plotting many columns at once.

```python
df3.iloc[:30].plot.area(alpha=0.4)
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
```

---

## Part 21 — When to Use What

| Tool | Best for |
|------|----------|
| **Pandas `.plot`** | Quick analysis, notebook work, debugging data, rapid feature checks |
| **Seaborn** | Nicer statistical plots, grouped plots, more polished visuals |
| **Matplotlib** | Fine control, publication-quality layouts, complex subplot construction |

> Use Pandas plotting for speed during EDA, then switch to Seaborn or Matplotlib when the figure needs to be shared or polished.

## Key Takeaways

- Pandas provides **fast plotting** directly from DataFrames and Series — ideal for exploratory analysis.
- The `.plot.xxx()` accessor syntax (`.plot.hist()`, `.plot.scatter()`, etc.) is the preferred interface.
- **Histograms**, **line plots**, **scatter plots**, **box plots**, **KDE plots**, and **hexbin plots** are especially useful.
- Matplotlib customisation still applies — pass `figsize`, `color`, `lw`, `alpha`, `cmap`, etc.
- Stylesheets (`plt.style.use('ggplot')`) can instantly improve visual appearance.
- Pandas plotting is convenient for speed, but **Seaborn** and **Matplotlib** remain important for more advanced visualisation work.
