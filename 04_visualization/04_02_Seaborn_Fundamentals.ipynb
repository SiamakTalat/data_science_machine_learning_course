## Part 1 — What Is Seaborn?

Seaborn is a **statistical visualization library** built on top of Matplotlib. It provides:
- Cleaner, more attractive default styles
- High-level statistical plotting functions (one-liners)
- Native integration with Pandas DataFrames

**When to use which:**

| Seaborn | Matplotlib |
|---------|------------|
| Statistical summaries, EDA | Full custom control |
| Works directly with DataFrames | Works with arrays |
| Faster to produce a readable chart | More verbose but more flexible |

> Seaborn and Matplotlib are complementary — you will often use both together.

## Part 2 — Setup

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline
```

```python
# Built-in example dataset: restaurant tipping data
tips = sns.load_dataset('tips')
tips.head()
```

**`tips` columns:** `total_bill`, `tip`, `sex`, `smoker`, `day`, `time`, `size`

> Seaborn ships with several built-in datasets. Load them with `sns.load_dataset('name')`.

## Part 3 — Distribution Plots

Distribution plots visualize how a single numeric variable is spread.

```python
# Histogram with KDE overlay
# Note: older tutorials use distplot() — histplot() is the modern equivalent
sns.histplot(tips['total_bill'], bins=30, kde=True)
plt.title('Histogram + KDE')
```

```python
# Histogram only (no KDE)
sns.histplot(tips['total_bill'], bins=30, kde=False)
plt.title('Histogram only')
```

```python
# KDE plot: smooth continuous density estimate
sns.kdeplot(tips['total_bill'])
plt.title('KDE Plot')
```

> A **KDE (Kernel Density Estimate)** smooths the histogram into a continuous curve — useful for comparing distributions without being sensitive to bin width.

```python
# Rug plot: a tick mark for every data point on the axis
sns.rugplot(tips['total_bill'])
plt.title('Rug Plot')
```

---

## Part 4 — Joint Plots

A joint plot combines a central bivariate chart with marginal distributions along each axis.

```python
# Default: scatter + histograms
sns.jointplot(data=tips, x='total_bill', y='tip')
```

```python
# kind='hex' — good for dense data
sns.jointplot(data=tips, x='total_bill', y='tip', kind='hex')
```

```python
# kind='reg' — scatter with linear regression
sns.jointplot(data=tips, x='total_bill', y='tip', kind='reg')
```

```python
# kind='kde' — smooth bivariate density
sns.jointplot(data=tips, x='total_bill', y='tip', kind='kde')
```

| `kind=` | Use when |
|---------|----------|
| `'scatter'` (default) | Moderate N, raw points |
| `'hex'` | Large N, dense overlapping points |
| `'reg'` | Want to see the linear trend |
| `'kde'` | Want a smooth density view |

## Part 5 — Pair Plots

`pairplot()` creates pairwise scatter plots for all numeric columns automatically.

```python
sns.pairplot(tips)
```

```python
# hue separates by a categorical variable using color
sns.pairplot(tips, hue='sex', palette='coolwarm')
```

> `pairplot()` is one of the most powerful single-line EDA tools — run it early on any new dataset.

## Part 6 — Categorical Plots

Categorical plots compare a numeric variable across categories.

```python
# Bar plot: shows mean (or other aggregate) per category
sns.barplot(data=tips, x='sex', y='total_bill')
plt.title('Bar Plot — mean total_bill by sex')
```

```python
# Custom aggregation: standard deviation instead of mean
sns.barplot(data=tips, x='sex', y='total_bill', estimator=np.std)
plt.title('Bar Plot — std of total_bill by sex')
```

```python
# Count plot: frequency count per category
sns.countplot(data=tips, x='sex')
plt.title('Count Plot — count by sex')
```

```python
# Box plot: median, IQR, and outliers
sns.boxplot(data=tips, x='day', y='total_bill')
plt.title('Box Plot')
```

```python
# Box plot with a second categorical split (hue)
sns.boxplot(data=tips, x='day', y='total_bill', hue='smoker')
plt.title('Box Plot with Hue')
```

```python
# Violin plot: distribution shape (KDE) + box plot summary
sns.violinplot(data=tips, x='day', y='total_bill')
plt.title('Violin Plot')
```

```python
# Violin with hue split
# Note: 'split=True' was the older API — in Seaborn >= 0.13, use 'dodge' or 'inner'
sns.violinplot(data=tips, x='day', y='total_bill', hue='sex')
plt.title('Violin Plot with Hue')
```

```python
# Strip plot: scatter of raw data points per category
sns.stripplot(data=tips, x='day', y='total_bill', jitter=True)
plt.title('Strip Plot (jittered)')
```

```python
# Swarm plot: non-overlapping scatter preserving distribution structure
sns.swarmplot(data=tips, x='day', y='total_bill')
plt.title('Swarm Plot')
```

```python
# Overlay: violin (density) + swarm (individual points)
sns.violinplot(data=tips, x='day', y='total_bill')
sns.swarmplot(data=tips, x='day', y='total_bill', color='black', size=3)
plt.title('Violin + Swarm Overlay')
```

**Choosing between categorical plots:**

| Plot | Shows |
|------|-------|
| `barplot` | Aggregate (mean/std) per category |
| `countplot` | Frequency count per category |
| `boxplot` | Median, quartiles, outliers |
| `violinplot` | Full distribution shape |
| `stripplot` | Raw data points (may overlap) |
| `swarmplot` | Raw data points (non-overlapping) |

```python
# catplot: unified high-level interface for any categorical plot
sns.catplot(data=tips, x='day', y='total_bill', kind='bar')
```

## Part 7 — Matrix Plots

Matrix plots require data already in **matrix form** — a correlation matrix or pivot table.

```python
# Heatmap on the tips correlation matrix
tc = tips.corr(numeric_only=True)

sns.heatmap(tc, annot=True, cmap='coolwarm')
plt.title('Tips Correlation Heatmap')
```

```python
# Heatmap on a pivot table: flights passengers by month and year
flights = sns.load_dataset('flights')

fp = flights.pivot_table(values='passengers', index='month', columns='year')
fp.head()
```

```python
sns.heatmap(fp, cmap='coolwarm', linewidths=0.5, linecolor='white')
plt.title('Flights Heatmap')
```

```python
# Clustermap: hierarchical clustering of rows and columns
sns.clustermap(fp, cmap='coolwarm', standard_scale=1)
# standard_scale=1 normalises each column to [0, 1]
```

> A **clustermap** reorders rows and columns by similarity — useful for spotting patterns not visible in fixed-order heatmaps.

## Part 8 — Grid-Based Plots

`PairGrid` and `FacetGrid` give granular control over multi-panel figures.

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
# PairGrid: manually control what goes in each position of the grid
g = sns.PairGrid(iris)
g.map_diag(sns.histplot)    # diagonal: distribution of each variable
g.map_upper(plt.scatter)    # upper triangle: scatter
g.map_lower(sns.kdeplot)    # lower triangle: KDE density
```

> **`pairplot()` vs `PairGrid`:** `pairplot()` is the quick wrapper; `PairGrid` lets you assign different plot types to the diagonal, upper, and lower triangles separately.

```python
# FacetGrid: repeat a plot across subsets of a dataset
g = sns.FacetGrid(tips, col='time', row='smoker')
g.map(sns.histplot, 'total_bill')
```

```python
# FacetGrid with scatter
g = sns.FacetGrid(tips, col='time', row='smoker')
g.map(plt.scatter, 'total_bill', 'tip')
```

## Part 9 — Regression Plots

`lmplot()` produces scatter plots with a linear regression fit. It also supports faceting.

```python
# Basic: scatter + regression line
sns.lmplot(data=tips, x='total_bill', y='tip')
```

```python
# Colored by category with custom markers
sns.lmplot(data=tips, x='total_bill', y='tip', hue='sex', markers=['o', 'v'])
```

```python
# Control scatter point size via scatter_kws (passed through to Matplotlib)
sns.lmplot(data=tips, x='total_bill', y='tip', hue='sex', scatter_kws={'s': 80})
```

```python
# Faceting: separate panel per day, colored by sex
sns.lmplot(data=tips, x='total_bill', y='tip', col='day', hue='sex')
```

```python
# Adjust panel size and aspect ratio
sns.lmplot(data=tips, x='total_bill', y='tip', col='day', hue='sex', aspect=0.8, height=5)
```

## Part 10 — Style, Context, and Palettes

```python
# Styles: 'white', 'ticks', 'darkgrid', 'whitegrid', 'dark'
sns.set_style('whitegrid')
sns.countplot(data=tips, x='sex')
plt.title('Style: whitegrid')
```

```python
# despine: remove top and right borders
sns.set_style('ticks')
sns.countplot(data=tips, x='sex')
sns.despine()
plt.title('Style: ticks + despine')
```

```python
# Figure size: control through Matplotlib before plotting
plt.figure(figsize=(10, 3))
sns.set_style('darkgrid')
sns.countplot(data=tips, x='sex')
plt.title('Custom size with darkgrid')
```

```python
# Context: scales text and element sizes for different audiences
# Options: 'paper', 'notebook' (default), 'talk', 'poster'
sns.set_context('talk')
sns.countplot(data=tips, x='sex')
plt.title('Context: talk')
```

```python
# Reset to notebook defaults for the rest of the notebook
sns.set_context('notebook')
sns.set_style('whitegrid')
```

```python
# Palette: controls the color scheme
sns.lmplot(data=tips, x='total_bill', y='tip', hue='sex', palette='coolwarm')
```

**Common palettes:** `'coolwarm'`, `'viridis'`, `'magma'`, `'Blues'`, `'Set1'`, `'Set2'`, `'rainbow'`

**Style guide:**
- `set_style()` → background and gridlines
- `set_context()` → text/element size for audience
- `palette=` → color scheme for categorical hue

---

## Key Takeaways

- Seaborn is built on Matplotlib — use them together; control size with `plt.figure(figsize=...)`.
- `histplot()` and `kdeplot()` replace the deprecated `distplot()` in modern Seaborn.
- `jointplot()` is the best single-function view of the relationship between two variables.
- `pairplot()` is the fastest way to explore all pairwise relationships in a dataset.
- Categorical plots: `barplot` → aggregates; `boxplot` → spread; `violinplot` → full distribution.
- Matrix plots (`heatmap`, `clustermap`) require the data to already be in matrix form.
- `PairGrid` and `FacetGrid` offer fine-grained control over multi-panel figures.
- `lmplot()` is the go-to for quick regression visualization, with built-in faceting support.
- `set_style()`, `set_context()`, and `palette=` control how figures look and scale.
