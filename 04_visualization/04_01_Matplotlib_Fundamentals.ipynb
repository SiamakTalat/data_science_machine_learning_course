## Part 2 — Setup

````python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
# %matplotlib inline renders plots directly inside notebook cells.
# If working in a plain .py script, call plt.show() at the end instead.
````

````python
# Sample data used throughout this notebook
x = np.linspace(0, 5, 11)
y = x ** 2

print("x:", x)
print("y:", y)
````

---

## Part 3 — Functional Interface (Quick Plots)

The functional interface uses `plt.*` functions directly. It is fast for quick exploration but has limited layout control.

````python
# Simplest possible plot
plt.plot(x, y)
````

````python
# Add labels and title
plt.plot(x, y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
````

````python
# Multiple plots with subplot(rows, cols, panel_number)
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r')   # red line

plt.subplot(1, 2, 2)
plt.plot(y, x, 'b')   # blue line
````

---

## Part 4 — Object-Oriented Interface (Preferred)

The OO interface explicitly creates `Figure` and `Axes` objects, giving full control over layout. **Use this for all serious work.**

```
Figure  →  the overall canvas
Axes    →  a single plotting area (one or more per figure)
```

````python
# Step 1: create a figure
fig = plt.figure()

# Step 2: add an axes — [left, bottom, width, height] as fractions of the figure (0-1)
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

# Step 3: plot on the axes object
axes.plot(x, y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
````

**`add_axes([left, bottom, width, height])`** — all four values are fractions of the figure canvas (0.0 to 1.0):
- `left` / `bottom` → position of the axes lower-left corner
- `width` / `height` → size of the axes

````python
# Inset plot: a main plot with a smaller secondary plot overlaid
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])   # main
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])   # inset

axes1.plot(x, y)
axes1.set_title('Larger Plot')

axes2.plot(y, x)
axes2.set_title('Smaller Plot')
````

---

## Part 5 — Subplots

`plt.subplots()` is the clean OO way to create a figure with multiple axes at once.

````python
# Single axes — equivalent to plt.figure() + add_axes but simpler
fig, ax = plt.subplots()
ax.plot(x, y)
````

````python
# Two side-by-side subplots
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x, y)
axes[0].set_title('First Plot')

axes[1].plot(y, x)
axes[1].set_title('Second Plot')

plt.tight_layout()   # prevents label/title overlap
````

````python
# Iterating through axes — useful for repeated plot patterns
fig, axes = plt.subplots(nrows=1, ncols=2)

for ax in axes:
    ax.plot(x, y)

plt.tight_layout()
````

> Always call `plt.tight_layout()` after multi-subplot figures to prevent overlapping titles and labels.

---

## Part 6 — Figure Size and DPI

````python
# figsize=(width_inches, height_inches)
fig = plt.figure(figsize=(8, 2))
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y)
````

````python
# figsize also works with subplots
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(8, 4))

axes[0].plot(x, y)
axes[1].plot(y, x)

plt.tight_layout()
````

---

## Part 7 — Saving Figures

````python
fig = plt.figure(figsize=(6, 4))
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y)
ax.set_title('Saved Figure')

# Save before calling plt.show() or at end of cell
# dpi controls resolution: 72 = screen, 150-300 = print quality
fig.savefig('my_picture.png', dpi=150, bbox_inches='tight')
````

**Supported formats:** `.png`, `.jpg`, `.svg`, `.pdf`

> Use `bbox_inches='tight'` to prevent labels from being clipped at the edge.

---

## Part 8 — Legends

Legends require a `label=` argument on each plot call.

````python
fig = plt.figure(figsize=(7, 4))
ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x**2, label='X Squared')
ax.plot(x, x**3, label='X Cubed')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Legend Example')

ax.legend()   # loc='best' or loc=0 lets Matplotlib choose the position
````

**Legend location options:**

| String | Integer code |
|--------|-------------|
| `'best'` | `0` |
| `'upper right'` | `1` |
| `'upper left'` | `2` |
| `'lower left'` | `3` |
| `'lower right'` | `4` |
| `'center'` | `10` |

## Part 9 — Colors, Linewidths, Linestyles, and Transparency

````python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(14, 4))

# Named color
axes[0].plot(x, y, color='purple')
axes[0].set_title('Named color')

# Hex color + transparency
axes[1].plot(x, y, color='#FF8C00', linewidth=4, alpha=0.5)
axes[1].set_title('Hex color + alpha')

# Linestyle
axes[2].plot(x, y, color='red', lw=2, ls='--')
axes[2].set_title('Dashed line')

plt.tight_layout()
````

**Linestyle shortcuts:**

| `ls=` | Appearance |
|-------|------------|
| `'-'` | Solid |
| `'--'` | Dashed |
| `'-.'` | Dash-dot |
| `':'` | Dotted |

**Key parameters:** `color`, `linewidth` / `lw`, `linestyle` / `ls`, `alpha` (0 = transparent, 1 = opaque)

---

## Part 10 — Markers

````python
# Simple marker
fig, ax = plt.subplots(figsize=(7, 4))
ax.plot(x, y, marker='o')
ax.set_title('Default circle markers')
````

````python
# Fully styled markers
fig, ax = plt.subplots(figsize=(7, 4))

ax.plot(
    x, y,
    color='blue',
    lw=1,
    marker='o',
    markersize=10,
    markerfacecolor='yellow',
    markeredgewidth=2,
    markeredgecolor='green'
)
ax.set_title('Custom marker styling')
````

**Common markers:** `'o'` circle, `'s'` square, `'^'` triangle, `'+'` plus, `'x'` cross, `'*'` star, `'D'` diamond

---

## Part 11 — Axis Limits

Use `set_xlim()` and `set_ylim()` to zoom into a region of interest.

````python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 4))

# Default limits
axes[0].plot(x, y, lw=2)
axes[0].set_title('Default limits')

# Zoomed in
axes[1].plot(x, y, lw=2, ls='--')
axes[1].set_xlim([0, 1])
axes[1].set_ylim([0, 2])
axes[1].set_title('Zoomed in')

plt.tight_layout()
````

## Part 12 — Mini Exercise: Putting It Together

````python
# Exercise: build a polished two-panel figure from scratch
# Left:  x vs x^2  — blue solid line, label 'Quadratic'
# Right: x vs x^3  — red dashed line, label 'Cubic'
# Add titles, labels, and legends to both panels

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 4))

axes[0].plot(x, x**2, color='blue', lw=2, label='Quadratic')
axes[0].set_title('Quadratic')
axes[0].set_xlabel('x')
axes[0].set_ylabel('x²')
axes[0].legend()

axes[1].plot(x, x**3, color='red', lw=2, ls='--', label='Cubic')
axes[1].set_title('Cubic')
axes[1].set_xlabel('x')
axes[1].set_ylabel('x³')
axes[1].legend()

plt.tight_layout()
````

---

## Key Takeaways

- Matplotlib's **functional API** (`plt.plot`) is fast but limited — use it for quick drafts.
- The **object-oriented API** (`fig, ax = plt.subplots()`) gives full control — prefer it for all real work.
- Every figure has `Figure` objects and `Axes` objects — these are the two central concepts.
- `plt.tight_layout()` is essential to prevent clipped or overlapping elements.
- Always pass `label=` to plot calls when you need a legend, then call `ax.legend()`.
- `figsize=(w, h)` in inches, `dpi` for resolution, `fig.savefig()` for export.
- Customisation follows the same pattern: `color`, `lw`, `ls`, `marker`, `alpha`, `set_xlim`, `set_ylim`.
