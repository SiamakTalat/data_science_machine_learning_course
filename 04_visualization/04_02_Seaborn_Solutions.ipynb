## Setup

```python
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline

titanic = sns.load_dataset('titanic')
```

---

## Exercise 1 — Joint Plot

```python
sns.jointplot(data=titanic, x='fare', y='age')
```

---

## Exercise 2 — Distribution Plot

```python
sns.histplot(titanic['fare'], bins=30, kde=False, color='red')
```

> **Version note:** Older Seaborn tutorials use `sns.distplot()`. This was deprecated in Seaborn 0.11 and removed in 0.12. The modern equivalent is `sns.histplot(kde=False)` for histograms and `sns.kdeplot()` for KDE curves.

---

## Exercise 3 — Box Plot

```python
sns.boxplot(data=titanic, x='class', y='age', palette='rainbow')
```

> **Reading a box plot:** the horizontal line inside the box is the **median**, the box spans the **IQR** (25th–75th percentile), the whiskers extend to ~1.5×IQR, and points beyond are **outliers**.

---

## Exercise 4 — Swarm Plot

```python
sns.swarmplot(data=titanic, x='class', y='age', palette='Set2')
```

> **Swarm vs strip:** `swarmplot` spreads points to avoid overlap while preserving the distribution shape. `stripplot` with `jitter=True` adds random noise. Swarm is cleaner for smaller datasets; strip is faster for large ones.

---

## Exercise 5 — Count Plot

```python
sns.countplot(data=titanic, x='sex')
```

---

## Exercise 6 — Heatmap

```python
plt.figure(figsize=(10, 7))
sns.heatmap(titanic.corr(numeric_only=True), cmap='coolwarm', annot=True, fmt='.2f')
plt.title('Titanic — Correlation Matrix')
```

> **Key arguments:** `annot=True` prints values in each cell, `fmt='.2f'` formats them to 2 decimal places, `cmap='coolwarm'` uses a diverging palette (red = positive, blue = negative correlation).

---

## Exercise 7 — FacetGrid

```python
g = sns.FacetGrid(titanic, col='sex')
g.map(sns.histplot, 'age')
```

```python
# Alternative using plt.hist (passes directly to Matplotlib)
g = sns.FacetGrid(titanic, col='sex')
g.map(plt.hist, 'age')
```

> **`FacetGrid` pattern:** create the grid with `col=` and/or `row=`, then call `.map(plot_function, 'column_name')`. Both Seaborn and Matplotlib functions can be passed.

---

## Reflection Answers

- **Joint vs pair plot:** joint plot shows two specific variables; pair plot shows all numeric pairs in the dataset.
- **Box vs swarm:** box plots summarize distribution compactly — use for large N or presentations. Swarm plots show every point — better for smaller datasets where individual values matter.
- **Heatmap requires matrix form:** `heatmap()` maps cell color to a numeric value — the data must already be arranged as rows × columns with a single value per cell.
- **Best for quick EDA:** `pairplot()` and `jointplot()` give the most information for the least code.

---

## Key Takeaways

- `histplot(kde=False)` is the current replacement for the deprecated `distplot()`.
- `boxplot` and `swarmplot` answer different questions — use both when the dataset is small enough.
- `heatmap(annot=True, fmt='.2f')` is the standard pattern for correlation matrices.
- `FacetGrid` requires: create grid → `.map(fn, 'column')` — two separate lines.
- Always set `figsize` before Seaborn calls that don't expose it directly.
