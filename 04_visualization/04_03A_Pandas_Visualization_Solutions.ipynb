## Setup

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df3 = pd.read_csv('df3')
df3.head()
```

---

## Exercise 1 — Scatter Plot

> Scatter plot of `B` vs `A` with custom width, point size, and color.

```python
df3.plot.scatter(x='A', y='B', figsize=(12, 3), s=50, c='red')
```

> `figsize=(12, 3)` creates a wide, short figure. `s=50` sets point size. `c='red'` sets point color.

---

## Exercise 2 — Histogram

> Histogram of column `A`.

```python
df3['A'].plot.hist()
```

---

## Exercise 3 — Styled Histogram

> Apply `ggplot` style, add bins and transparency.

```python
plt.style.use('ggplot')
df3['A'].plot.hist(bins=25, alpha=0.5)
```

> `plt.style.use('ggplot')` applies the R-inspired ggplot theme. `alpha=0.5` adds transparency so overlapping bars are visible.

---

## Exercise 4 — Box Plot

> Box plot comparing only columns `A` and `B`.

```python
df3[['A', 'B']].plot.box()
```

> Select the columns first with `df3[['A', 'B']]` to limit the box plot to only those two.

---

## Exercise 5 — KDE Plot

> KDE plot for column `D`.

```python
df3['D'].plot.kde()
```

---

## Exercise 6 — Styled KDE Plot

> KDE with increased line width and dashed style.

```python
df3['D'].plot.kde(lw=3, ls='--')
```

> `lw=3` sets linewidth. `ls='--'` sets the linestyle to dashed. These are Matplotlib keyword arguments passed through by Pandas.

---

## Exercise 7 — Area Plot

> Area plot of all columns for the first 30 rows.

```python
df3.iloc[:30].plot.area(alpha=0.4)
```

> `.iloc[:30]` selects the first 30 rows by integer position. `alpha=0.4` adds transparency so overlapping areas are visible.

---

## Exercise 8 — Bonus: Legend Outside

> Move the legend outside the figure to the right.

```python
df3.iloc[:30].plot.area(alpha=0.4)
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
```

> `bbox_to_anchor=(1.0, 0.5)` places the legend anchor at the right edge (x=1.0) and vertically centred (y=0.5). `loc='center left'` aligns the legend's left edge to that anchor point.
